### Creando los datasets y ajustando sus distribuciones.

Crear 3 datasets con respuestas de clientes a una encuesta de satisfacción de producto. Las encuestas tienen una escala de 1 a 7 (discreta), donde nota 1 es muy baja satisfacción y la nota 7 es muy alta satisfacción con el producto.
Los datasets son los siguientes:

1. **Capel**: crear un conjunto con 800 encuestas de satisfacción, que representan la historia del último trimestre de Capel. La satisfacción en promedio fue de **4.50** puntos y tuvo una distribición similar a la de la siguiente figura para que la use de referencia.

2. **Competencia**: crear un dataset con 500 encuestas de satisfacción, que representan la historia del último trimestre de la principal marca de la competencia. 3. **Le Pac**: crear un dataset con 320 encuestas de satisfacción, que representan la satisfacción de clientes de este nuevo producto, estas encuestas fueron realizadas durante 1 mes. La satisfacción en promedio fue de **6.15** puntos y tuvo una distribición similar a la de la siguiente figura para que la use de referencia.

```python
# Tu código aquí ...
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Función para crear un dataset con distribución truncada
def crear_dataset(n, media, desviacion, min_val=1, max_val=7):
    # Generar datos con distribución normal
    datos = np.random.normal(media, desviacion, n)
    # Truncar los datos para que estén en el rango [min_val, max_val]
    datos_truncados = np.clip(datos, min_val, max_val)
    return datos_truncados

# Parámetros para cada dataset
params = {
    'Capel': {'n': 800, 'media': 4.50, 'desviacion': 1.0},
    'Competencia': {'n': 500, 'media': 6.10, 'desviacion': 0.8},
    'Le Pac': {'n': 320, 'media': 6.15, 'desviacion': 0.75}
}

# Crear datasets
datasets = {}
for nombre, param in params.items():
    datasets[nombre] = crear_dataset(param['n'], param['media'], param['desviacion'])

# Convertir a DataFrame para facilitar el manejo
df_capel = pd.DataFrame(datasets['Capel'], columns=['Satisfacción'])
df_competencia = pd.DataFrame(datasets['Competencia'], columns=['Satisfacción'])
df_le_pac = pd.DataFrame(datasets['Le Pac'], columns=['Satisfacción'])
```

## <font color='green'>**Actividad 2**</font>
### Estudiando los datasets

Para cada uno de los datasets creados se pide:

1. Graficar el correspondiente histograma y su función de masa de probabilidad y/o distribución de probabilidad según corresponda.

2. Calcular la media y desviación estandar de la muestra.

3. Calcular el porcentaje de notas 6 y 7. Este indicador en Marketing es conocido como el **Top 2 Box Score**, que corresponde a una proproción de las personas que han manifestado satisfacción por el producto/servicio.

4. Calcular el intervalo de confianza para la media y proporción Top2Box. Ambas con un nivel de significancia alpha del 0.05.




```python
# Tu código aquí ...
# Función para calcular el intervalo de confianza
def intervalo_confianza_media(data, alpha=0.05):
    n = len(data)
    media = np.mean(data)
    std = np.std(data, ddof=1)  # Desviación estándar muestral
    error_estandar = std / np.sqrt(n)

    # Valor crítico t
    t_critico = stats.t.ppf(1 - alpha/2, df=n-1)

    margen_error = t_critico * error_estandar
    return (media - margen_error, media + margen_error)

# Función para calcular el intervalo de confianza para proporciones
def intervalo_confianza_proporcion(successes, n, alpha=0.05):
    p_hat = successes / n
    error = np.sqrt((p_hat * (1 - p_hat)) / n)

    z_critico = stats.norm.ppf(1 - alpha/2)

    margen_error = z_critico * error
    return (p_hat - margen_error, p_hat + margen_error)

# Crear un DataFrame con los datasets generados previamente
datasets = {
    'Capel': df_capel['Satisfacción'],
    'Competencia': df_competencia['Satisfacción'],
    'Le Pac': df_le_pac['Satisfacción']
}

# Analizar cada dataset
for nombre, data in datasets.items():
    # Graficar histograma y KDE
    plt.figure(figsize=(10, 5))
    sns.histplot(data, bins=7, kde=True, stat="density", color='skyblue', edgecolor='black')

    # Función de densidad teórica
    mu = np.mean(data)
    std = np.std(data, ddof=1)

    x = np.linspace(1, 7, 100)
    plt.plot(x, stats.norm.pdf(x, mu, std), 'r-', lw=2, label='PDF Teórica')

    plt.title(f'Histograma y PDF - {nombre}')
    plt.xlabel('Satisfacción')
    plt.ylabel('Densidad')
    plt.xlim(1, 7)
    plt.legend()
    plt.grid()
    plt.show()

    # Calcular media y desviación estándar
    media = np.mean(data)
    desviacion_std = np.std(data, ddof=1)

    print(f'{nombre} - Media: {media:.2f}, Desviación Estándar: {desviacion_std:.2f}')

    # Calcular porcentaje de notas 6 y 7 (Top 2 Box Score)
    top_2_box_count = np.sum((data == 6) | (data == 7))
    top_2_box_percentage = (top_2_box_count / len(data)) * 100

    print(f'{nombre} - Top 2 Box Score: {top_2_box_percentage:.2f}%')

    # Calcular intervalos de confianza para la media y proporción Top 2 Box
    ic_media = intervalo_confianza_media(data)

    print(f'{nombre} - Intervalo de Confianza para la Media: ({ic_media[0]:.2f}, {ic_media[1]:.2f})')

    ic_proporcion_top_2_box = intervalo_confianza_proporcion(top_2_box_count, len(data))

    print(f'{nombre} - Intervalo de Confianza para Proporción Top 2 Box: ({ic_proporcion_top_2_box[0]*100:.2f}%, {ic_proporcion_top_2_box[1]*100:.2f}%)\n')
```

## <font color='green'>**Actividad 3**</font>
### Creando nuestro test de hipótesis

Desarrollar un test de hipótesis de 1 muestra para al menos 1 de los 3 datasets. Aquí Ud. puede formular la hipótesis a probar. Por ejemplo, puede ser contra un target definido por la industria o retail.



```python
# Tu código aquí ...

data_capel = df_capel['Satisfacción']
# Parámetros del test
mu_0 = 5  # Media hipotética (target definido por la industria)
alpha = 0.05  # Nivel de significancia

# Calcular la media y desviación estándar muestral
media_muestral = np.mean(data_capel)
desviacion_std = np.std(data_capel, ddof=1)  # Desviación estándar muestral
n = len(data_capel)  # Tamaño de la muestra

# Calcular el estadístico t
t_statistic = (media_muestral - mu_0) / (desviacion_std / np.sqrt(n))

# Grados de libertad
df = n - 1

# Calcular el valor p (prueba unilateral)
p_value = stats.t.cdf(t_statistic, df)

# Resultados del test
print(f'Media Muestral: {media_muestral:.2f}')
print(f'Desviación Estándar: {desviacion_std:.2f}')
print(f'Estadístico t: {t_statistic:.4f}')
print(f'Valor p: {p_value:.4f}')

# Decisión sobre la hipótesis
if p_value < alpha:
    print("Rechazamos la hipótesis nula. La satisfacción media es menor que 5.")
else:
    print("No rechazamos la hipótesis nula. No hay evidencia suficiente para afirmar que la satisfacción media es menor que 5.")
```

## <font color='green'>**Actividad 4**</font>
### Validación de algunos test de hipótesis

Desarrollar los siguientes test de hipótesis de 2 muestras para:

1. Verificar si la satisfacción de Le Pac es mayor a la del producto Capel.

2. Verificar si la satisfacción de Le Pac es mayor a la Competencia.

```python
# Tu código aquí ...
data_le_pac = df_le_pac['Satisfacción']
data_capel = df_capel['Satisfacción']
data_competencia = df_competencia['Satisfacción']

# Función para realizar el test t de dos muestras
def test_hipotesis(data1, data2, alpha=0.05):
    # Calcular medias y desviaciones estándar
    media1 = np.mean(data1)
    media2 = np.mean(data2)
    std1 = np.std(data1, ddof=1)
    std2 = np.std(data2, ddof=1)

    # Tamaños de muestra
    n1 = len(data1)
    n2 = len(data2)

    # Calcular el estadístico t para la prueba unilateral
    t_statistic = (media1 - media2) / np.sqrt((std1**2 / n1) + (std2**2 / n2))

    # Grados de libertad
    df = n1 + n2 - 2

    # Calcular valor p para la prueba unilateral
    p_value = 1 - stats.t.cdf(t_statistic, df)

    return media1, media2, t_statistic, p_value

# Test 1: Satisfacción Le Pac vs Capel
print("Test de Hipótesis: Satisfacción Le Pac vs Capel")
media_le_pac, media_capel, t_stat_lepac_capel, p_value_lepac_capel = test_hipotesis(data_le_pac, data_capel)

print(f'Media Le Pac: {media_le_pac:.2f}, Media Capel: {media_capel:.2f}')
print(f'Estadístico t: {t_stat_lepac_capel:.4f}, Valor p: {p_value_lepac_capel:.4f}')

if p_value_lepac_capel < 0.05:
    print("Rechazamos la hipótesis nula. La satisfacción media de Le Pac es mayor que la de Capel.")
else:
    print("No rechazamos la hipótesis nula. No hay evidencia suficiente para afirmar que la satisfacción media de Le Pac es mayor que la de Capel.")

print("\n" + "="*50 + "\n")

# Test 2: Satisfacción Le Pac vs Competencia
print("Test de Hipótesis: Satisfacción Le Pac vs Competencia")
media_competencia, _, t_stat_lepac_competencia, p_value_lepac_competencia = test_hipotesis(data_le_pac, data_competencia)

print(f'Media Competencia: {media_competencia:.2f}, Media Le Pac: {media_le_pac:.2f}')
print(f'Estadístico t: {t_stat_lepac_competencia:.4f}, Valor p: {p_value_lepac_competencia:.4f}')

if p_value_lepac_competencia < 0.05:
    print("Rechazamos la hipótesis nula. La satisfacción media de Le Pac es mayor que la de Competencia.")
else:
    print("No rechazamos la hipótesis nula. No hay evidencia suficiente para afirmar que la satisfacción media de Le Pac es mayor que la de Competencia.")
```
