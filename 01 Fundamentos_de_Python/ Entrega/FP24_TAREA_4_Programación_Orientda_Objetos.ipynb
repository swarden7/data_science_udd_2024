## <font color='green'>Tarea 1: Challenging</font>
Cree una clase Esfera que acepte un radio en la instanciación y tenga  métodos de volumen y área de superficie.

Tip:

* Utiliza la función `round( )` para ajustar los decimales de salida

```python
# Tu código aquí ...

class Esfera():
    pi = 3.14159

    def __init__(self, radio):
        self.radio = radio

    def volumen(self):
        """ Devuelve el volumen de la esfera"""
        return round(((4/3) * self.pi * self.radio**3), 2)

    def area_superficie(self):
        """ Devuelve el área de la esfera"""
        return round((4 * self.pi * self.radio**2), 2)




```

```python
s = Esfera(1)
```

```python
s.area_superficie()
```

```python
s.volumen()
```

## <font color='green'>Tarea 2: Super Challenging</font>

**Este ejercicio lo desarrollaremos en varios pasos.**
1. Crea una clase llamada *GuessingGame*.
2. La clase no necesita ningún argumento proporcionado por el usuario para la instanciación.
3. En el método `.__init__()` (el constructor), haz que la clase establezca un atributo `self.rand_choice` asignándole un número entero aleatorio entre 0-10.
4. Crea un método en la clase llamado *reset_random* que vuelva a generar un número aleatorio (también entre 0-10) para el atributo `self.rand_choice`
5. Crea un método llamado *guess* (adivinar) que use la función `input()` para aceptar un número del usuario. Este método debe imprimir una declaración que indique si la suposición del usuario fue correcta o no.

Bonus track: agrega una lógica en el método *guess* que informe al usuario cuando su número está por arriba o por debajo del número a adivinar.

Tip:
* En el punto 3 use `import random`
* Te dejamos el esqueleto de la clase

```python
# Tu código aquí ...

import random
# Esqueleto de la clase
class GuessingGame():


    def __init__(self):
        import random
        self.rand_choice = random.randint(0,10)
        print("Se ha instanciado el juego")


    def reset_random(self):
        self.rand_choice = random.randint(0,11)
        print("Se ha cambiado el número")


    def guess(self):
        while True:
            try:
                self.numero_usuario = input("Ingrese un número entre 0 y 10. Para salir ingrese x.")
                #en dos líneas distintas para permitir que el usuario haga el input
                self.numero_usuario = int(self.numero_usuario)

                if self.numero_usuario == self.rand_choice:
                    print(f"¡Adivinaste! El número efectivamente era el {self.rand_choice}")


                elif (self.numero_usuario < 0) or (self.numero_usuario > 10):
                    print("El número debe estar entre 0 y 10... Intenta nuevamente")


                else:
                    if self.numero_usuario < self.rand_choice:
                        print("El número es un poco más grande... Intenta nuevamente.")
                    else:
                        print("El número es un poco más pequeño... Intenta nuevamente.")

            except ValueError:
                if self.numero_usuario == 'x':
                    print("Saliendo...")
                    break

                print(f"Sólo se aceptan números enteros entre 0 y 10...")



# Solución
```

```python
g = GuessingGame()
```

```python
g.rand_choice
```

```python
g.reset_random()
```

```python
g.guess()
```

### Modelando un zoológico
Crea una jerarquía de clases para modelar un zoológico. Debe haber una clase Animal con atributos comunes a todos los animales. Luego, crea clases para diferentes tipos de animales (por ejemplo, Mamífero, Reptil, Ave), cada una de las cuales hereda de Animal. Cada una de estas clases debe tener atributos y métodos específicos de su tipo. Por último, crea clases para animales específicos (por ejemplo, León, Cocodrilo, Águila) que heredan de sus respectivas clases de tipo de animal. Utiliza el polimorfismo para crear un método en la clase Animal que describa el sonido que hace el animal, pero implementa este método de manera diferente en cada clase de animal específico.

```python
# Tu código aquí ...

class Animal():

    def __init__(self, nombre):
        self.zoo = "Zoológico de Elochito"
        self.filo = 'Vertebrado'
        self.animal = 'animal'

    def habla(self):
        if self.sonido:
            print(f"¡Hola! Soy un {self.animal}, me llamo {self.nombre} y hago {self.sonido}")

        else:
            print(f"¡Hola! Soy un {self.animal}, me llamo {self.nombre} y no hago sonido.")


class Mamifero(Animal):
    clase = 'Mamíferos'

    def __init__(self, nombre):
        # Atributos de instancia estándar,
        # se pueden cambiar en las clases que hereden de esta clase, patas, pelo, escamas, de_tierra, de_agua
        self.patas = 4
        self.pelo = True
        self.escamas = False
        self.de_tierra = True
        self.de_agua = True
        self.vuela = False
        self.nombre = 'Animal'
        self.sonido = None


class Reptil(Animal):
    clase = 'Reptiles'

    def __init__(self, nombre, animal, clase):
        # Atributos de instancia estándar,
        # se pueden cambiar en las clases que hereden de esta
        self.animal = animal
        self.patas = 4
        self.pelo = False
        self.escamas = True
        self.de_tierra = True
        self.de_agua = False
        self.vuela = True


class Pez(Animal):
    clase = 'Peces'

    def __init__(self, nombre, animal):
        self.animal = animal
        self.patas = 0
        self.pelo = False
        self.escamas = True
        self.de_tierra = False
        self.de_agua = True
        self.vuela = False

class Ave(Animal):
    clase = 'Aves'

    def __init__(self, nombre, animal):
        self.animal = animal
        self.patas = 2
        self.pelo = False
        self.escamas = True
        self.de_tierra = False
        self.de_agua = False
        self.vuela = True


class Puma(Mamifero):
    def __init__(self, nombre):
        self.animal = 'Puma'
        self.especie = 'Puma concolor'
        self.sonido = 'grrrraaauuuuu!'
        self.nombre = nombre
        self.clase = super().clase

class Delfin(Mamifero):
    def __init__(self, nombre):
        self.animal = 'Delfín Chileno'
        self.nombre = nombre
        self.especie = 'Cephalorhynchus eutropia'
        self.sonido = 'ik ik ik ik ik!'
        self.clase = super().clase

        #se modifican algunos atributos de las clases de las cuales hereda
        self.patas = 0
        self.pelo = False
        self.de_tierra = False
        self.de_agua = True

class Condor(Ave):
    def __init__(self, nombre):
        self.animal = 'Cóndor'
        self.nombre = nombre
        self.especie = 'Vultur gryphus'
        self.sonido = 'wiiiuu wiuuu lallalalaaa!'
        self.clase = super().clase

class Pinguino(Ave):
    def __init__(self, nombre):
        self.animal = 'Pingüino de Humboldt'
        self.nombre = nombre
        self.especie = 'Spheniscus humboldti'
        self.sonido = 'tuuuuuuu aaargh tuuuuuu aaaaargh!'
        self.de_tierra = True
        self.de_agua = True
        self.clase = super().clase
class Atun(Pez):
    def __init__(self, nombre):
        self.animal = 'Atún'
        self.nombre = nombre
        self.especie = 'Thunnus thynnus'
        self.sonido = 'glup glup glup!'
        self.clase = super().clase

# Instanciamos animales

espumita = Puma('Espumita')
natu = Atun('Natu')
condorito = Condor('Condorito')
pingu = Pinguino('Pingu')
delprincipio = Delfin('Delprincipio')


# Así hablan

espumita.habla()
natu.habla()
condorito.habla()
pingu.habla()
delprincipio.habla()

# Heredan atributos

print("\nEstos animales heredaron de las siguientes clases' \n")
print('Pumita ->', espumita.clase)
print('Natu ->', natu.clase)
print('Cóndor ->', condorito.clase)
print('Pingu ->', pingu.clase)
print('Delprincipio ->',delprincipio.clase)

print('\nAhora le quitaremos el sonido al Cóndor')
condorito.sonido = None
condorito.habla()
```

### Modelando una biblioteca
Crea una jerarquía de clases para modelar una biblioteca. Debe haber una clase Libro con atributos comunes a todos los libros. Luego, crea clases para diferentes tipos de libros (por ejemplo, Novela, CienciaFiccion, Biografia), cada una de las cuales hereda de Libro. Cada una de estas clases debe tener atributos y métodos específicos de su tipo. Utiliza el polimorfismo para crear un método en la clase Libro que describa el tipo de lectura que proporciona el libro, pero implementa este método de manera diferente en cada clase de libro específico.

```python
# Tu código aqui ...

class Libro():
    biblioteca = 'Biblioteca Municipal'

    def __init__(self, titulo, autores:list):
        self.titulo = titulo
        self.autores = autores
        self.genero = 'No categorizado'

    def info(self):
        # Muestra los autores dependiendo si la liste tiene uno o más elementos
        if len(self.autores) == 1:
            print(f"Título\t\t: {self.titulo}")
            print(f"Autor\t\t: {self.autores[0]}")
            print(f"Género\t\t: {self.genero}")

        elif len(self.autores) > 1:
            print(f"Título\t\t: {self.titulo}")
            print(f"Autores\t\t: {self.autores[0]}")
            for autor in self.autores[1:]:
                print(f"\t\t  {autor}")

            print(f"Género\t\t: {self.genero}")



class Infantil(Libro):

    def __init__(self, titulo:str, autores:list, edades:tuple):
        super().__init__(titulo, autores)
        self.genero = 'Infantil'
        self.edades = f"Desde los {edades[0]} a los {edades[1]} años"

    def info(self):
        super().info() #method override
        print(f"Edades\t\t: {self.edades}")

class Ciencia(Libro):

    def __init__(self, titulo:str, autores:list, area='No definida'):
        super().__init__(titulo, autores)
        self.genero = 'Ciencia'
        self.area = area

    def info(self):
        super().info()
        print(f"Área\t\t: {self.area}")

class Novela(Libro):
    def __init__(self, titulo:str, autores:list, subgenero='No definido'):
        super().__init__(titulo, autores)
        self.genero = 'Novela'
        self.subgenero = subgenero

    def info(self):
        super().info()
        print(f"Subgénero\t: {self.subgenero}")


```

```python
print('*'*80)
python = Ciencia('Python Basics: A Practical Introduction to Python 3', ['David Amos', 'Dan Bader', 'Joanna Jablonski', 'Fletcher Heisler'], 'Programación' )
python.info()
print('*'*80)
principito = Infantil('El Principito', ['Antoine de Saint-Exupéry'], (10, 99))
principito.info()
print('*'*80)
gatsby = Novela('The Great Gatsby', ['F. Scott Fitzgerald'], 'Great american novel')
gatsby.info()
print('*'*80)
```
