## <font color='blue'>**Definiciones**</font>

Una hip√≥tesis estad√≠stica es una afirmaci√≥n o proposici√≥n respecto a alguna caracter√≠stica de una poblaci√≥n, generalmente fundamentada sobre un par√°metro de la misma. Contrastar una hip√≥tesis es comparar las predicciones con la realidad que observamos ocurrida en una muestra. Si dentro del margen de error que estamos dispuestos a admitir, hay coincidencia, aceptaremos la hip√≥tesis y en caso contrario la rechazaremos.

1. La hip√≥tesis emitida se suele designar por¬†$ùëØ_ùüé$¬†y se llama¬†**Hip√≥tesis nula**. Lo de ‚Äúnula‚Äù viene de que partimos del supuesto que las diferencias entre el valor verdadero del par√°metro y su valor hipot√©tico, en realidad, no son tales sino debidas al azar, es decir, no hay diferencia entre ambos, o dicho de otra forma la diferencia es nula. Las siguientes son hip√≥tesis nulas t√≠picas:
<br>
<br>
$H_0: \quad \mu = 37$<br>
$H_0: \quad p = 0.5$<br>
$H_0: \quad \sigma = 15$
<br>
<br>

2. La hip√≥tesis contraria se designa por¬†$ùëØ_ùüè$¬†¬†y se llama¬†**Hip√≥tesis alternativa**. La hip√≥tesis alternativa es la afirmaci√≥n de que el par√°metro tiene un valor que difiere de la hip√≥tesis nula. Los s√≠mbolos que emplea la hip√≥tesis alternativa son: $\lt$, $\gt$ o bien $\ne$.
<br>
<br>
$$
\begin{align*}
\text{Medias}: &\quad H_1:\mu \gt 37 \quad H_1: \mu \lt 37 &\quad H_1: \mu \neq 37\\
\text{Proporciones}: &\quad H_1:p \gt 0.5 \quad H_1: p \lt 0.5 &\quad H_1: p \neq 0.5 \\
\text{Desviaci√≥n est√°ndar}: &\quad \underbrace{H_1:\sigma \gt 15 \quad H_1: \sigma \lt 15 }_{Unilaterales\\(una \, cola)}&\quad \underbrace{H_1: \sigma \neq 15}_{Bilaterales\\(dos \, colas)}
\end{align*}
$$
<br>
<br>

### __Contraste de hip√≥tesis__
Se formula la hip√≥tesis sobre la poblaci√≥n (e.g., $\mu$, $p$, $\sigma$, etc.).  Entonces las conclusiones sobre la validez de la hip√≥tesis se basar√°n en la informaci√≥n de una muestra (e.g.,. $\bar{x}$, $\hat{p}$, $S$, etc).

La realizaci√≥n de un contraste de hip√≥tesis __no consiste en__ dicidir cu√°l de las dos hip√≥tesis ($H_0$ o $H_1$) es m√°s creible, sino en decidir si la muestra proporciona o no suficiente evidencia para descartar $H_0$. Mientras no tengamos suficiente evidencia para aceptar $H_1$ tendremos que creernos que $H_0$ es cierta.

## <font color='blue'>**Metodolog√≠a de Pruebas de Hip√≥tesis**</font>

El proceso de prueba de hip√≥tesis sigue una secuencia estructurada de pasos:
1. Formulaci√≥n de Hip√≥tesis: Establecer $H_0$ y $H_1$ en t√©rminos del problema a investigar.
2.	Selecci√≥n del Nivel de Significancia ( $\alpha$ ): Determinar la probabilidad m√°xima de cometer un error Tipo I (rechazar $H_0$ cuando es verdadera).
3.	Elecci√≥n del Estad√≠stico de Prueba: Seleccionar el estad√≠stico adecuado (e.g., $Z$, $t$) basado en el tama√±o muestral y el conocimiento de par√°metros poblacionales.
4. Determinaci√≥n de la Regla de Decisi√≥n: Definir los valores cr√≠ticos o la regi√≥n de rechazo para $H_0$.
    - M√©todo Cl√°sico: Comparar el valor del estad√≠stico de prueba con los valores cr√≠ticos. Si el estad√≠stico cae en la regi√≥n cr√≠tica, se rechaza $H_0$.
    - M√©todo del $Valor-P$: Calcular la probabilidad de observar un resultado tan extremo como el obtenido, suponiendo que $H_0$ es verdadera. Si el $Valor-P$ es menor o igual a $ \alpha $, se rechaza $H_0$.
5.	Interpretaci√≥n de Resultados: Comparar el estad√≠stico calculado con los valores cr√≠ticos o analizar el $Valor-P$ para tomar una decisi√≥n.

## <font color='blue'>**Pruebas de Hip√≥tesis de Una Muestra**</font>

### __Media Poblacional__
- **$ \sigma $ conocido (Prueba $Z$):**
   - Se utiliza cuando la desviaci√≥n est√°ndar poblacional ($ \sigma $) es conocida.
   - $ \sigma $ desconocido si el tama√±o de la muestra es suficientemente grande ($n > 30$) $^{(1)}$
   - El estad√≠stico $Z$ se calcula mediante la f√≥rmula:
   $$
   Z = \frac{\bar{X} - \mu_0}{\sigma / \sqrt{n}}
   $$
   - **C√°lculo del $p$-value:**
     - Para una prueba unilateral:
       $$
       \text{p-value} = P(Z > |z|)
       $$
     - Para una prueba bilateral:
       $$
       \text{p-value} = 2 \times P(Z > |z|)
       $$
   donde:
   - $ \bar{X} $: Media muestral.
   - $ \mu_0 $: Media poblacional bajo $ H_0 $.
   - $ n $: Tama√±o de la muestra.
   - $ \sigma $: Desviaci√≥n est√°ndar poblacional conocida.

- **$ \sigma $ desconocido (Prueba $t$):**
   - Se aplica cuando $ \sigma $ es desconocida y se utiliza la desviaci√≥n est√°ndar muestral ($ S $).
   - El estad√≠stico $t$ se calcula como:
   $$
   t = \frac{\bar{X} - \mu_0}{S / \sqrt{n}}
   $$
   - **C√°lculo del $p$-value:**
     - Para una prueba unilateral:
       $$
       \text{p-value} = P(t > |t|)
       $$
     - Para una prueba bilateral:
       $$
       \text{p-value} = 2 \times P(t > |t|)
       $$
   donde:
   - $ S $: Desviaci√≥n est√°ndar muestral.
   - $ t $: Sigue una distribuci√≥n t de Student con $ n-1 $ grados de libertad.

Nota:

(1) Aunque t√©cnicamente es v√°lido usar la prueba $Z$ en estas condiciones, muchos estad√≠sticos prefieren usar la prueba $t$ en lugar de $Z$, ya que la prueba $t$ est√° espec√≠ficamente dise√±ada para manejar la incertidumbre de $\sigma$, independientemente del tama√±o de la muestra. Sin embargo, para $n > 30$, las distribuciones $Z$ y $t$ son casi id√©nticas, por lo que los resultados ser√°n muy similares.

### __Proporci√≥n Poblacional (Prueba $Z$)__
- Se utiliza para contrastar una proporci√≥n muestral ($ \hat{p} $) con una proporci√≥n poblacional hipot√©tica ($ p_0 $).
- El estad√≠stico $Z$ se calcula como:
$$
Z = \frac{\hat{p} - p_0}{\sqrt{\frac{p_0(1-p_0)}{n}}}
$$
- **C√°lculo del $p$-value:**
  - Para una prueba unilateral:
    $$
    \text{p-value} = P(Z > |z|)
    $$
  - Para una prueba bilateral:
    $$
    \text{p-value} = 2 \times P(Z > |z|)
    $$
donde:
   - $ \hat{p} $: Proporci√≥n muestral.
   - $ p_0 $: Proporci√≥n poblacional bajo $ H_0 $.
   - $ n $: Tama√±o de la muestra.
   

## Pasos para la Prueba de Hip√≥tesis: Proporciones

### 1. Plantear la hip√≥tesis nula y la hip√≥tesis alternativa
- **Hip√≥tesis Nula (H‚ÇÄ):** Generalmente establece que la proporci√≥n poblacional $p$ es igual a un valor hipot√©tico $p_0$:  
  $H‚ÇÄ: p = p_0$

- **Hip√≥tesis Alternativa (H‚ÇÅ):** Establece que la proporci√≥n puede ser diferente, mayor o menor que $p_0$:  
  $H‚ÇÅ: p \neq p_0$ (o $p > p_0$ o $p < p_0$).

---

### 2. Muestra y proporci√≥n muestral
- Se extrae una muestra de tama√±o $n$.  
- Se calcula la proporci√≥n muestral $\hat{p}$ como:  
  $\hat{p} = \frac{\text{N√∫mero de √©xitos en la muestra}}{n}$

---

### 3. Distribuci√≥n de $\hat{p}$ bajo $H‚ÇÄ$
- Si $n$ es grande, por el Teorema del L√≠mite Central, la proporci√≥n muestral $\hat{p}$ se distribuye aproximadamente como una normal con:  
  - Media: $p_0$  
  - Varianza: $\frac{p_0 (1 - p_0)}{n}$

  En notaci√≥n:  
  $\hat{p} \sim N\left(p_0, \frac{p_0 (1 - p_0)}{n}\right)$ (aprox. para $n$ grande).

---

### 4. Estandarizaci√≥n
Para contrastar hip√≥tesis sobre $\hat{p}$, se construye un estad√≠stico que sigue, aproximadamente, una distribuci√≥n normal est√°ndar ($N(0, 1)$).

- **Media te√≥rica bajo $H‚ÇÄ$:** $p_0$  
- **Desviaci√≥n est√°ndar te√≥rica bajo $H‚ÇÄ$:** $\sqrt{\frac{p_0 (1 - p_0)}{n}}$

El estad√≠stico de prueba $Z$ se define como:  
$Z = \frac{\hat{p} - p_0}{\sqrt{\frac{p_0 (1 - p_0)}{n}}}$

Este estad√≠stico $Z$ sigue aproximadamente una distribuci√≥n normal est√°ndar:  
$Z \sim N(0, 1)$ (bajo $H‚ÇÄ$).

---

### 5. Regla de Decisi√≥n
- **P-valor (valor-p):** Con el valor observado de $Z$, se calcula la probabilidad (bajo la curva normal est√°ndar) de observar un valor tan extremo o m√°s extremo que el observado.  
- **Nivel de significancia $\alpha$:** Si el valor-p $<$ $\alpha$, se rechaza $H‚ÇÄ$; de lo contrario, no se rechaza $H‚ÇÄ$.

---

### 6. Resumen de la F√≥rmula Final
La f√≥rmula del estad√≠stico $Z$ para contrastar una proporci√≥n muestral $\hat{p}$ contra la proporci√≥n hipot√©tica $p_0$ es:  
$Z = \frac{\hat{p} - p_0}{\sqrt{\frac{p_0 (1 - p_0)}{n}}}$

Este estad√≠stico se interpreta bajo la distribuci√≥n normal est√°ndar para tomar decisiones sobre $H‚ÇÄ$.

**Punto clave:** La derivaci√≥n se basa en la aproximaci√≥n normal de la distribuci√≥n de $\hat{p}$. Cuanto mayor sea $n$, m√°s justificada estar√° dicha aproximaci√≥n.


### Paso a paso para obtener la varianza de la proporci√≥n muestral $\hat{p}$

#### **1. Definici√≥n de la proporci√≥n muestral**
La proporci√≥n muestral se define como:
$ \hat{p} = \frac{\text{N√∫mero de √©xitos en la muestra}}{n} $
donde:
- $\hat{p}$: Proporci√≥n muestral.
- $n$: Tama√±o de la muestra.

El n√∫mero de √©xitos en la muestra sigue una distribuci√≥n binomial:
$ X \sim \text{Binomial}(n, p_0) $
donde:
- $X$: N√∫mero de √©xitos en la muestra.
- $p_0$: Proporci√≥n poblacional bajo la hip√≥tesis nula $H_0$.

#### **2. Propiedades de la distribuci√≥n binomial**
Para una variable aleatoria $X$ que sigue una distribuci√≥n binomial $\text{Binomial}(n, p_0)$, las propiedades relevantes son:
- **Media**: $E[X] = n p_0$.
- **Varianza**: $\text{Var}(X) = n p_0 (1 - p_0)$.

#### **3. Relaci√≥n entre $X$ y $\hat{p}$**
La proporci√≥n muestral $\hat{p}$ se obtiene dividiendo $X$ por $n$:
$ \hat{p} = \frac{X}{n} $

Por propiedades de la varianza, si una variable aleatoria $Y$ se divide por una constante $c$, su varianza cambia seg√∫n:
$ \text{Var}\left(\frac{Y}{c}\right) = \frac{\text{Var}(Y)}{c^2} $

Aplicando esta regla a $\hat{p}$:
$ \text{Var}(\hat{p}) = \frac{\text{Var}(X)}{n^2} $

#### **4. Sustituci√≥n de la varianza de $X$**
Sabemos que:
$ \text{Var}(X) = n p_0 (1 - p_0) $

Sustituyendo en la ecuaci√≥n de $\text{Var}(\hat{p})$:
$ \text{Var}(\hat{p}) = \frac{n p_0 (1 - p_0)}{n^2} $

#### **5. Simplificaci√≥n**
Simplificando la ecuaci√≥n:
$ \text{Var}(\hat{p}) = \frac{p_0 (1 - p_0)}{n} $

#### **6. Desviaci√≥n est√°ndar**
La desviaci√≥n est√°ndar, que es la ra√≠z cuadrada de la varianza, es:
$ \text{Desviaci√≥n est√°ndar} = \sqrt{\frac{p_0 (1 - p_0)}{n}} $

#### **7. Conclusi√≥n**
La varianza de la proporci√≥n muestral $\hat{p}$ bajo la hip√≥tesis nula $H_0$ es:
$ \text{Var}(\hat{p}) = \frac{p_0 (1 - p_0)}{n} $
Esta f√≥rmula se utiliza para calcular el estad√≠stico de prueba $Z$ en pruebas de hip√≥tesis sobre proporciones.



## <font color='blue'>**Estad√≠stico de prueba**</font>
El **estad√≠stico de prueba**, es un valor que se utiliza para tomar la decisi√≥n sobre la hip√≥tesis nula y se calcula convirtiendo el **estad√≠stico muestral** (tales como la media muestral $\bar{x}$, o la proporci√≥n muestral $\hat{p}$) en una puntuaci√≥n (tales como los $Z$ o $t$ _scores_), bajo el supuesto que la hip√≥tesis nula es verdadera. Este tipo de tests, utilizar√° puntuaciones $Z$ o $t$ dependiendo del tama√±o de la muestra y de si $\sigma$, (desviaci√≥n est√°ndar poblacional) es conocida.

### <font color='blue'>**Nivel de significaci√≥n ($\alpha$)**</font>

El **nivel de significaci√≥n ($\alpha$)** es una probabilidad prefijada que define el l√≠mite aceptable para cometer un error Tipo I (rechazar la hip√≥tesis nula $H_0$ cuando en realidad es verdadera). En el contexto de la inferencia estad√≠stica, representa la proporci√≥n de la distribuci√≥n bajo $H_0$ que se encuentra fuera del intervalo de confianza y en la regi√≥n de rechazo.

#### **Prop√≥sito del Nivel de Significaci√≥n**
- Establece un criterio objetivo para decidir si se rechaza o no $H_0$.
- Determina el umbral para considerar un resultado como estad√≠sticamente significativo.
- Define la probabilidad m√°xima aceptada de obtener un resultado extremo debido al azar, incluso cuando $H_0$ es verdadera.

#### **Valores Comunes de $\alpha$**
- Los valores m√°s utilizados son:
  - $\alpha = 0.05$ (5%): Se tolera un error en 5 de cada 100 pruebas.
  - $\alpha = 0.01$ (1%): Se tolera un error en 1 de cada 100 pruebas.
- La elecci√≥n del valor de $\alpha$ depende del contexto y las consecuencias del error Tipo I en el problema analizado.

#### **Interpretaci√≥n**
Un nivel de significaci√≥n del $5\%$ significa que, en promedio, si $H_0$ es verdadera, se espera que el resultado caiga en la regi√≥n de rechazo 5 de cada 100 veces simplemente por azar. Este l√≠mite no garantiza la certeza absoluta de las conclusiones, sino que establece un marco probabil√≠stico para la toma de decisiones estad√≠sticas.

#### **Importancia**
El nivel de significaci√≥n es una herramienta clave en la inferencia estad√≠stica, ya que equilibra el riesgo de rechazar una hip√≥tesis verdadera mientras permite detectar efectos significativos con una confianza predefinida.

### <font color='blue'>**Prueba $Z$**</font>

La **prueba $Z$** es una herramienta estad√≠stica param√©trica utilizada para evaluar la significancia de las diferencias entre medias o proporciones, bajo el supuesto de que la desviaci√≥n est√°ndar poblacional ($\sigma$) es conocida. Es apropiada para muestras amplias (generalmente $n > 30$) y datos independientes, es decir, que no est√°n correlacionados entre s√≠. Este enfoque se basa en la distribuci√≥n normal est√°ndar y utiliza la media muestral, la desviaci√≥n est√°ndar conocida y el tama√±o de la muestra para calcular el estad√≠stico $Z$.

El estad√≠stico $Z$ se calcula como:
$$
Z = \frac{\bar{X} - \mu_0}{\sigma / \sqrt{n}}
$$

Una vez obtenido el valor de $Z$, se compara con un valor cr√≠tico, determinado a partir de la distribuci√≥n normal est√°ndar y del nivel de significancia elegido ($\alpha$). La hip√≥tesis nula ($H_0$), que generalmente establece que no existe una diferencia significativa entre la media muestral y la media poblacional, ser√° rechazada si $Z$ excede el valor cr√≠tico en la regi√≥n de rechazo.

### <font color='blue'>**T-Test**</font>


__Test-T__ es una prueba en la que el estad√≠stico utilizado tiene una distribuci√≥n __t de Student__ si la hip√≥tesis nula es cierta. Se aplica cuando la poblaci√≥n estudiada sigue una distribuci√≥n normal, pero el tama√±o muestral es demasiado peque√±o como para que el estad√≠stico en el que est√° basada la inferencia est√© normalmente distribuido, utiliz√°ndose una estimaci√≥n de la desviaci√≥n t√≠pica en lugar del valor real.

### <font color='blue'>**Valor-P ($p$-value)**</font>

El **$p$-value** es el nivel de significaci√≥n m√°s peque√±o para el cual los datos muestrales proporcionar√≠an suficiente evidencia para rechazar la hip√≥tesis nula ($H_0$). En t√©rminos pr√°cticos, el $p$-value nos ayuda a evaluar si un resultado observado es producto del azar o si es estad√≠sticamente significativo. Es una medida probabil√≠stica que indica la compatibilidad de los datos con $H_0$.

#### **Caracter√≠sticas del $p$-value**
- Es la probabilidad (calculada bajo la suposici√≥n de que $H_0$ es verdadera) de obtener un resultado igual o m√°s extremo que el observado.
- Depende exclusivamente de los datos de la muestra.
- Indica el grado de apoyo que los datos proporcionan a $H_0$:
  - **Si el $p$-value es menor que el nivel de significaci√≥n ($\alpha$):**
    - Se considera que los datos proporcionan suficiente evidencia contra $H_0$.
    - El resultado es **estad√≠sticamente significativo**, y se **rechaza** $H_0$.
  - **Si el $p$-value es mayor o igual que el nivel de significaci√≥n ($\alpha$):**
    - No hay suficiente evidencia para rechazar $H_0$.
    - Se **acepta provisionalmente** que $H_0$ es consistente con los datos.
- Como una probabilidad, el $p$-value siempre se encuentra en el rango [0, 1].

#### **Interpretaci√≥n**
- **$p$-value bajo:** Indica que es poco probable observar los datos obtenidos si $H_0$ fuera verdadera, lo que sugiere que $H_0$ debe rechazarse.
- **$p$-value alto:** Indica que los datos son consistentes con $H_0$ y no hay suficiente evidencia para rechazarla.

#### **C√°lculo del $p$-value**
El c√°lculo del $p$-value depende del estad√≠stico de prueba y del tipo de prueba (unilateral o bilateral). Generalmente, el procedimiento sigue los siguientes pasos:
1. Formular $H_0$ y $H_1$.
2. Calcular el estad√≠stico de prueba (e.g., $Z$ o $t$).
3. Determinar la probabilidad asociada al valor del estad√≠stico:
   - Para pruebas unilaterales:
     $$
     \text{$p$-value} = P(\text{estad√≠stico} > \text{observado})
     $$
   - Para pruebas bilaterales:
     $$
     \text{$p$-value} = 2 \times P(\text{estad√≠stico} > |\text{observado}|)
     $$
4. Comparar el $p$-value con el nivel de significaci√≥n ($\alpha$).

### Hip√≥tesis en el ejemplo del c√°lculo del estad√≠stico Z

En el ejemplo del c√≥digo proporcionado, las hip√≥tesis nula (H0) y alternativa (H1) se definen de la siguiente manera:

---

#### **Hip√≥tesis Nula (H0)**
- La media de la poblaci√≥n (mu) es igual a la media poblacional asumida (mu0 = 100).
- Matem√°ticamente:
  H0: mu = 100
- Esto implica que no hay una diferencia estad√≠sticamente significativa entre la media de la muestra (muestra_media) y la media poblacional hipot√©tica (mu0).

---

#### **Hip√≥tesis Alternativa (H1)**
- La media de la poblaci√≥n (mu) es diferente de la media poblacional asumida (mu0 = 100).
- Matem√°ticamente (prueba bilateral):
  H1: mu ‚â† 100
- Esto indica que existe una diferencia significativa entre la media de la muestra y la media poblacional hipot√©tica.

---

### **Interpretaci√≥n de las hip√≥tesis**
- **Si rechazamos H0:**
  - La evidencia estad√≠stica sugiere que la media poblacional no es igual a 100; es decir, la diferencia observada en la muestra no se debe al azar.
- **Si no rechazamos H0:**
  - No hay suficiente evidencia estad√≠stica para concluir que la media poblacional sea diferente de 100.

---

### **Contexto del c√≥digo**
- **H0:** La muestra podr√≠a haber venido de una poblaci√≥n cuya media es 100.
- **H1:** La muestra proviene de una poblaci√≥n cuya media es significativamente distinta de 100.


```
# Importaci√≥n de la funci√≥n norm de scipy.stats para c√°lculos de la distribuci√≥n normal
from scipy.stats import norm
import numpy as np  # Asegurar la disponibilidad de la funci√≥n np.sqrt

# Definici√≥n de los datos de la muestra
sample_mean = 105  # Media muestral (promedio observado en la muestra) ###### Probar con 108
population_mean = 100  # Media poblacional bajo la hip√≥tesis nula (H0)
sigma = 15  # Desviaci√≥n est√°ndar poblacional conocida
n = 30  # Tama√±o de la muestra
alpha = 0.05  # Nivel de significancia (umbral para tomar decisiones)

# C√°lculo del estad√≠stico Z
# F√≥rmula: Z = (media_muestral - media_poblacional) / (sigma / sqrt(n))
z_score = (sample_mean - population_mean) / (sigma / np.sqrt(n))

# C√°lculo del Valor-P para una prueba bilateral
# Valor-P = 2 * P(Z > |z_score|) bajo la hip√≥tesis nula
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Decisi√≥n basada en el Valor-P
# Si el Valor-P es menor que el nivel de significancia (alpha), rechazamos H0
if p_value < alpha:
    decision = "Rechazar H0 (estad√≠sticamente significativo)"
else:
    decision = "No rechazar H0 (no estad√≠sticamente significativo)"

# Impresi√≥n de los resultados
print(f"Estad√≠stico Z: {z_score:.2f}")  # Muestra el estad√≠stico Z con dos decimales
print(f"Valor-P: {p_value:.6f}")       # Muestra el Valor-P con seis decimales
print(f"Decisi√≥n: {decision}")        # Muestra la decisi√≥n final basada en el Valor-P
```



### <font color='blue'>**Criterios de decisi√≥n**</font>

El procedimiento convencional de prueba de hip√≥tesis requiere que __siempre probemos la hip√≥tesis nula__, de manera que nuestra conclusi√≥n inicial siempre ser√° una de las siguientes:

- Rechazo la hip√≥tesis nula.
- No rechazo la hip√≥tesis nula.

La decisi√≥n de rechazar o no una hip√≥tesis nula, se puede realizar a trav√©s de dos m√©todos; el __M√©todo cl√°sico__ o por __Valor P__:

#### <font color='blue'>__1. M√©todo cl√°sico__</font>

- __Rechazo $H_0$__  si el estad√≠stico de prueba ($Z$ o $t$) cae dentro de la Regi√≥n cr√≠tica.

- __No Rechazo $H_0$__  si el estad√≠stico de prueba __NO__ cae dentro de la Regi√≥n cr√≠tica.

#### **Regiones Cr√≠ticas o de Rechazo**

#### **1.1 Prueba Bilateral**
- **Hip√≥tesis:**
  - $H_0: \mu = \mu_0$
  - $H_1: \mu \neq \mu_0$
- **Regi√≥n de rechazo (con $\sigma$ conocida):**
  $$
  Z \leq -Z_{\alpha/2} \quad \text{o} \quad Z \geq Z_{\alpha/2}
  $$
- **Regi√≥n de rechazo (con $\sigma$ desconocida):**
  $$
  t \leq -t_{n-1, \alpha/2} \quad \text{o} \quad t \geq t_{n-1, \alpha/2}
  $$

#### **1.2 Prueba Unilateral (Izquierda)**
- **Hip√≥tesis:**
  - $H_0: \mu \geq \mu_0$
  - $H_1: \mu < \mu_0$
- **Regi√≥n de rechazo (con $\sigma$ conocida):**
  $$
  Z \leq -Z_{\alpha}
  $$
- **Regi√≥n de rechazo (con $\sigma$ desconocida):**
  $$
  t \leq -t_{n-1, \alpha}
  $$

#### **1.3 Prueba Unilateral (Derecha)**
- **Hip√≥tesis:**
  - $H_0: \mu \leq \mu_0$
  - $H_1: \mu > \mu_0$
- **Regi√≥n de rechazo (con $\sigma$ conocida):**
  $$
  Z \geq Z_{1-\alpha}
  $$
- **Regi√≥n de rechazo (con $\sigma$ desconocida):**
  $$
  t \geq t_{n-1, 1-\alpha}
  $$

### **Notas:**
1. Los valores cr√≠ticos para la distribuci√≥n $t$ (cuando $\sigma$ es desconocida) se calculan usando `t.ppf` en lugar de `norm.ppf`.
2. Estos ejemplos est√°n dise√±ados para datos con $\sigma$ conocida. Si $\sigma$ es desconocida, se reemplaza con la desviaci√≥n est√°ndar muestral ($S$) y se usa la distribuci√≥n $t$.

```
# Importaci√≥n de bibliotecas necesarias
from scipy.stats import norm, t  # norm para distribuci√≥n normal, t para pruebas adicionales si es necesario.

# Par√°metros iniciales
alpha = 0.05  # Nivel de significancia (probabilidad m√°xima de cometer error tipo I).
sample_mean = 105  # Media observada en la muestra.
population_mean = 100  # Media poblacional bajo la hip√≥tesis nula (H0).
sigma = 15  # Desviaci√≥n est√°ndar poblacional conocida.
n = 30  # Tama√±o de la muestra.

# C√°lculo del estad√≠stico Z
# F√≥rmula: Z = (media_muestral - media_hipot√©tica) / (sigma / sqrt(n))
z_score = (sample_mean - population_mean) / (sigma / (n**0.5))

# Prueba bilateral: Considera ambas colas (valores extremos en ambas direcciones).
z_critical_bilateral = norm.ppf(1 - alpha / 2)  # Valor cr√≠tico para prueba bilateral.
reject_H0_bilateral = abs(z_score) >= z_critical_bilateral  # Decisi√≥n: comparar |Z| con el valor cr√≠tico.
# Imprime resultados de la prueba bilateral
print(f"Prueba Bilateral - Estad√≠stico Z: {z_score:.2f}, Valor cr√≠tico: {z_critical_bilateral:.2f}")
print(f"Rechazar H0 (bilateral): {reject_H0_bilateral}\n")

# Prueba unilateral izquierda: Considera solo la cola izquierda (valores menores).
z_critical_left = norm.ppf(alpha)  # Valor cr√≠tico para cola izquierda.
reject_H0_left = z_score <= z_critical_left  # Decisi√≥n: comparar Z con el valor cr√≠tico.
# Imprime resultados de la prueba unilateral izquierda
print(f"Prueba Unilateral Izquierda - Estad√≠stico Z: {z_score:.2f}, Valor cr√≠tico: {z_critical_left:.2f}")
print(f"Rechazar H0 (izquierda): {reject_H0_left}\n")

# Prueba unilateral derecha: Considera solo la cola derecha (valores mayores).
z_critical_right = norm.ppf(1 - alpha)  # Valor cr√≠tico para cola derecha.
reject_H0_right = z_score >= z_critical_right  # Decisi√≥n: comparar Z con el valor cr√≠tico.
# Imprime resultados de la prueba unilateral derecha
print(f"Prueba Unilateral Derecha - Estad√≠stico Z: {z_score:.2f}, Valor cr√≠tico: {z_critical_right:.2f}")
print(f"Rechazar H0 (derecha): {reject_H0_right}")
```

```
import matplotlib.pyplot as plt

# Datos generales
alpha = 0.05
x = np.linspace(-4, 4, 500)
y = norm.pdf(x, 0, 1)

# Estad√≠stico Z de prueba
z_prueba = 1.5  # Puedes modificar este valor para probar distintos Z

# Configuraci√≥n de los valores cr√≠ticos
z_critical_bilateral = norm.ppf(1 - alpha / 2)
z_critical_left = norm.ppf(alpha)
z_critical_right = norm.ppf(1 - alpha)

# Configuraci√≥n del gr√°fico
fig, axes = plt.subplots(3, 1, figsize=(8, 12))

# Gr√°fico 1: Prueba Bilateral
axes[0].plot(x, y, label="Distribuci√≥n Normal", color="blue")
axes[0].fill_between(x, y, where=(x <= -z_critical_bilateral) | (x >= z_critical_bilateral), color="red", alpha=0.5, label="Regi√≥n de rechazo")
axes[0].axvline(-z_critical_bilateral, color="black", linestyle="--", label=f"Z = -{z_critical_bilateral:.2f}")
axes[0].axvline(z_critical_bilateral, color="black", linestyle="--", label=f"Z = {z_critical_bilateral:.2f}")
axes[0].axvline(z_prueba, color="green", linestyle=":", label=f"Z de prueba = {z_prueba:.2f}")
axes[0].set_title("Prueba Bilateral")
axes[0].legend()
axes[0].set_xlabel("Z")
axes[0].set_ylabel("Densidad")

# Gr√°fico 2: Prueba Unilateral (Izquierda)
axes[1].plot(x, y, label="Distribuci√≥n Normal", color="blue")
axes[1].fill_between(x, y, where=(x <= z_critical_left), color="red", alpha=0.5, label="Regi√≥n de rechazo")
axes[1].axvline(z_critical_left, color="black", linestyle="--", label=f"Z = {z_critical_left:.2f}")
axes[1].axvline(z_prueba, color="green", linestyle=":", label=f"Z de prueba = {z_prueba:.2f}")
axes[1].set_title("Prueba Unilateral (Izquierda)")
axes[1].legend()
axes[1].set_xlabel("Z")
axes[1].set_ylabel("Densidad")

# Gr√°fico 3: Prueba Unilateral (Derecha)
axes[2].plot(x, y, label="Distribuci√≥n Normal", color="blue")
axes[2].fill_between(x, y, where=(x >= z_critical_right), color="red", alpha=0.5, label="Regi√≥n de rechazo")
axes[2].axvline(z_critical_right, color="black", linestyle="--", label=f"Z = {z_critical_right:.2f}")
axes[2].axvline(z_prueba, color="green", linestyle=":", label=f"Z de prueba = {z_prueba:.2f}")
axes[2].set_title("Prueba Unilateral (Derecha)")
axes[2].legend()
axes[2].set_xlabel("Z")
axes[2].set_ylabel("Densidad")

# Ajustar dise√±o
plt.tight_layout()
plt.show()
```

### <font color='blue'>**Ejemplo: Hip√≥tesis para la media  ($\sigma$ conocido)**</font>


Una manada de 1.500 novillos fue alimentada con un grano especial rico en prote√≠nas durante un mes. Se pes√≥ una muestra aleatoria de 29 y hab√≠an ganado un promedio de 7.7 libras. Si la desviaci√≥n est√°ndar de la ganancia de peso para todo el reba√±o es 7.1, pruebe la hip√≥tesis de que la ganancia de peso promedio por novillo durante el mes fue de m√°s de 5 libras.

1. Hipotesis nula, $H_0:\mu = 5$
2. Hip√≥tesis alternativa, $H_1:\mu > 5$

Vamos a calcular $Z$:

$$Z = \frac{7.7 - 5}{\frac{7.1}{\sqrt{29}}} = 2.049$$

Ahora en Python:



```
from scipy.stats import norm

# Datos del problema
sample_mean = 7.7  # Media muestral
population_mean = 5  # Media bajo H0
sigma = 7.1  # Desviaci√≥n est√°ndar poblacional
n = 29  # Tama√±o de la muestra
alpha = 0.05  # Nivel de significancia

# C√°lculo del estad√≠stico Z
z_score = (sample_mean - population_mean) / (sigma / np.sqrt(n))

# Valor cr√≠tico para prueba unilateral (derecha)
z_critical = norm.ppf(1 - alpha)

# Decisi√≥n basada en el estad√≠stico Z
reject_H0 = z_score > z_critical

# Resultados
print(f"Estad√≠stico Z: {z_score:.3f}")
print(f"Valor cr√≠tico Z: {z_critical:.3f}")
if reject_H0:
    print("Se rechaza la hip√≥tesis nula (H0): hay evidencia de que la ganancia promedio es mayor a 5 libras.")
else:
    print("No se rechaza la hip√≥tesis nula (H0): no hay evidencia suficiente para afirmar que la ganancia promedio es mayor a 5 libras.")
```

Si pedimos un valor de 5% de significaci√≥n, el valor del $z$ cr√≠tico es de 1.645, por lo tanto como nuestro $z$ es mayor que dicho valor (2.048), se rechaza la hipotesis nula, con lo cual hubo ganancia de peso, estimada con 5% de significancia.

### Hip√≥tesis para la media  ($\sigma$ conocido)

Se desea contrastar con un nivel de significaci√≥n del 5 % la hip√≥tesis de que ‚Äúla talla media de los hombres de 18 o m√°s a√±os de un pa√≠s es igual a 180‚Äù. Suponiendo que la desviaci√≥n est√°ndar de las tallas en la poblaci√≥n es 4, contraste dicha hip√≥tesis frente a la alternativa de que es distinta.

Los datos constituyen una muestra de $n=15$ hombres seleccionados al azar, cuyas alturas son: 167, 167, 168, 168, 168, 169, 171, 172, 173, 175, 175, 175, 177, 182, 195.

$ùëã$: Altura de los hombres mayores o igual a 18 a√±os<br>
$X \sim (\mu, 4)$

1. Calcular estadisticos muestra
2. Definir hip√≥tesis
3. Definir tipo de prueba a aplicar
4. Calcular Z cr√≠tico
5. Calcular estad√≠stico de prueba
6. Definir regla decisi√≥n
7. Calcular valor-p
8. Interprete y concluya
9. Grafique la solcui√≥n

```
# Tu c√≥digo aqu√≠ ...

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos del problema
muestra = [167, 167, 168, 168, 168, 169, 171, 172, 173, 175, 175, 175, 177, 182, 195]
mu_0 = 180  # Media hipot√©tica
sigma = 4   # Desviaci√≥n est√°ndar conocida
n = len(muestra)  # Tama√±o de la muestra
nivel_significacion = 0.05

# C√°lculo de estad√≠sticos muestrales
media_muestral = np.mean(muestra)

# Definir hip√≥tesis
# H0: mu = 180
# H1: mu != 180

# Tipo de prueba: Prueba bilateral

# C√°lculo del valor cr√≠tico (Z cr√≠tico)
z_critico = stats.norm.ppf(1 - nivel_significacion / 2)

# C√°lculo del estad√≠stico de prueba (Z observado)
z_observado = (media_muestral - mu_0) / (sigma / np.sqrt(n))

# Regla de decisi√≥n
# Rechazar H0 si |Z observado| > Z cr√≠tico

# C√°lculo del valor-p
valor_p = 2 * (1 - stats.norm.cdf(abs(z_observado)))

# Resultados e interpretaci√≥n
print(f"Media muestral: {media_muestral:.2f}")
print(f"Estad√≠stico Z observado: {z_observado:.2f}")
print(f"Valor cr√≠tico Z: {z_critico:.2f}")
print(f"Valor-p: {valor_p:.4f}")

if abs(z_observado) > z_critico:
    print("Se rechaza la hip√≥tesis nula H0.")
else:
    print("No se rechaza la hip√≥tesis nula H0.")

# Gr√°fica de la soluci√≥n
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Distribuci√≥n Z", color="blue")

# Regiones de rechazo
plt.fill_between(x, y, where=(x <= -z_critico), color="red", alpha=0.5, label="Regi√≥n de rechazo")
plt.fill_between(x, y, where=(x >= z_critico), color="red", alpha=0.5)

# Estad√≠stico observado
plt.axvline(z_observado, color="green", linestyle="--", label=f"Z observado = {z_observado:.2f}")

# Configuraci√≥n de la gr√°fica
plt.title("Contraste de hip√≥tesis para la media (prueba Z bilateral)")
plt.xlabel("Z")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid()
plt.show()
```

### <font color='blue'>**Ejemplo: Hip√≥tesis para la media  ($\sigma$ desconocido)**</font>
### Rentabilidad fondos de inversi√≥n

Sea X la variable ‚Äùrentabilidad de cierto tipo de fondos de inversi√≥n tras una apreciaci√≥n fuerte del Euro con respecto al D√≥lar‚Äù. Se considera que la media de esta variable es 15. Un economista afirma que dicha rentabilidad media ha variado, por lo que lleva a cabo un estudio en las condiciones rese√±adas anteriormente, sobre una muestra de 9 fondos cuya media muestral resulta ser de 15.308 y cuya varianza muestral es 0.193.

1. Especificando las hip√≥tesis necesarias, contrastar la afirmaci√≥n del economista al 5%.

2. A partir del resultado de punto 1, razonar si el intervalo de confianza para la media al 95% contendr√° o no al valor 15.

```
# 1.  calcular estadisticos muestra
xbar = 15.308
S2 = 0.193 # varianza muestral
S = np.sqrt(S2) # desv. std muestral insesgada (asumimos)
n = 9

print(f'Xbarra:         {xbar:4.2f}')
print(f'S:              {S:4.2f}')
print(f'Tama√±o muestra: {n}')


# 2. Definir hip√≥tesis a contrastar (prueba bilateral - 2 colas)
#    H0:   Mu = 15
#    H1:   Mu ‚â† 15

# 3. como n < 30 y no se conoce Sigma, aplicaremos T Student
alpha = 0.05
mu = 15

# 4. calcular valor cr√≠tico de T
#    n-1 grados de libertad, donde n corresponde al n√∫mero de la muestra y 2 colas
Tcritico = stats.t.ppf(1-(alpha/2), df=(n-1))
print(f'T cr√≠tico: {Tcritico:4.2f}')


# 5. calcular Estadistico de prueba t
t =  (xbar - mu) / (S / np.sqrt(n))
print(f't: {t:4.2f}')

# 6. Define Regla de decision
#    no rechaza H0 si: -Tcritico <= t <= Tcritico
#    rechazar H0 si:  t < -Tcritico  o  t > Tcritico


# 7. calcular valor-p
pvalue = 1 - (stats.t.cdf(t, df=(n - 1)))
print(f'Valor-p: {pvalue*2:4.2f}') #2 colas
```

Conclusi√≥n:  

Como $t$ de prueba se encuentra entre $+- Tcritico$ , no se rechaza la $H_0$. No existe evidencia para afirmar que la rentabilidad ha cambiado.

Al observar el valor-p , este es mayor a $\alpha=0,05$. Por tanto no se rechaza $H_0$ ya que 7 de cada 100 veces que la hip√≥tesis nula sea cierta, ser√° rechazada por el azar de la muestra.

```
import numpy as np
from scipy.stats import t

# Datos proporcionados
dulces = np.array([0.751, 0.841, 0.856, 0.799, 0.966, 0.859, 0.857, 0.942,
                   0.873, 0.809, 0.890, 0.878, 0.905])

# Estad√≠sticos de muestra
n = len(dulces)
media_muestral = np.mean(dulces)
desv_std_muestral = np.std(dulces, ddof=1)  # ddof=1 para desviaci√≥n est√°ndar muestral

# Hip√≥tesis
# H0: Œº <= 0.8535 (Los M&M tienen un peso promedio menor o igual al indicado)
# H1: Œº > 0.8535 (Los M&M tienen un peso promedio mayor al indicado)

media_hipotetica = 0.8535
nivel_significancia = 0.05

# Estad√≠stico de prueba (t de Student)
t_stat = (media_muestral - media_hipotetica) / (desv_std_muestral / np.sqrt(n))

# Valor cr√≠tico (t cr√≠tico para una cola)
t_critico = t.ppf(1 - nivel_significancia, df=n-1)

# Valor-P
valor_p = 1 - t.cdf(t_stat, df=n-1)

# Resultados
print(f"Estad√≠sticos de la muestra:")
print(f"- Tama√±o de la muestra (n): {n}")
print(f"- Media muestral: {media_muestral:.4f}")
print(f"- Desviaci√≥n est√°ndar muestral: {desv_std_muestral:.4f}\n")

print(f"Prueba de hip√≥tesis:")
print(f"- Estad√≠stico de prueba (t): {t_stat:.4f}")
print(f"- Valor cr√≠tico (t cr√≠tico): {t_critico:.4f}")
print(f"- Valor-P: {valor_p:.4f}\n")

# Regla de decisi√≥n
if t_stat > t_critico:
    print("Conclusi√≥n: Se rechaza la hip√≥tesis nula (H0).")
    print("Los datos sugieren que los M&M tienen un peso promedio mayor al indicado en la etiqueta.")
else:
    print("Conclusi√≥n: No se puede rechazar la hip√≥tesis nula (H0).")
    print("No hay suficiente evidencia para afirmar que los M&M tienen un peso promedio mayor al indicado.")
```

### <font color='blue'>**Decisiones en Pruebas de Hip√≥tesis y los Errores Asociados**</font>

Cuando se realiza una prueba de hip√≥tesis, el objetivo es llegar a una conclusi√≥n basada en los datos disponibles: rechazar la hip√≥tesis nula ($H_0$) o no rechazarla. Sin embargo, incluso si la prueba se realiza correctamente, las conclusiones no son absolutamente ciertas debido a la naturaleza probabil√≠stica de las pruebas estad√≠sticas. Toda prueba de hip√≥tesis conlleva un riesgo de error.

Los errores asociados a las decisiones en pruebas de hip√≥tesis se dividen en dos categor√≠as principales: el **Error Tipo I** y el **Error Tipo II**. Estos errores tienen un impacto significativo en la validez de los resultados y est√°n relacionados de manera inversa. Por lo tanto, es fundamental analizar las consecuencias de cada tipo de error en el contexto del problema antes de establecer los niveles de significancia y potencia de la prueba.

---

### <font color='blue'>**Error de Tipo I**</font>

Un **Error Tipo I** ocurre cuando se **rechaza la hip√≥tesis nula ($H_0$) siendo verdadera**. Este error se mide por el nivel de significancia ($\alpha$), el cual es definido a priori por el investigador.

- **Interpretaci√≥n de $\alpha$:**
  - Si $\alpha = 0.05$, significa que se acepta un 5% de probabilidad de cometer un error al rechazar una hip√≥tesis nula que es cierta.
  - En t√©rminos pr√°cticos, se est√° dispuesto a aceptar un peque√±o margen de error para detectar efectos significativos.

- **Consecuencias del Error Tipo I:**
  - Concluir que existe un efecto o diferencia cuando en realidad no la hay.
  - Puede llevar a decisiones innecesarias o incorrectas, especialmente en campos como la medicina o las pol√≠ticas p√∫blicas.

- **Mitigaci√≥n:**
  - Reducir $\alpha$ (e.g., de 0.05 a 0.01) disminuye la probabilidad de cometer un error de tipo I, pero aumenta el riesgo de cometer un error de tipo II ($\beta$).

$$
\text{Nivel de significancia} = \alpha
$$

---

### <font color='blue'>**Error de Tipo II**</font>

Un **Error Tipo II** ocurre cuando se **falla al rechazar la hip√≥tesis nula ($H_0$) siendo falsa**. La probabilidad de cometer este error se denota como $\beta$.

- **Interpretaci√≥n de $\beta$:**
  - Si $\beta = 0.20$, significa que hay un 20% de probabilidad de no detectar una diferencia real cuando esta existe.

- **Consecuencias del Error Tipo II:**
  - No detectar un efecto o diferencia que realmente est√° presente.
  - Puede llevar a pasar por alto descubrimientos importantes, como efectos cl√≠nicos significativos o tendencias cr√≠ticas en investigaciones cient√≠ficas.

- **Relaci√≥n con la Potencia de la Prueba:**
  - La potencia de la prueba ($1 - \beta$) es la probabilidad de rechazar $H_0$ cuando es falsa. Un valor t√≠pico deseado es $1 - \beta = 0.80$ o mayor.
  - La potencia de la prueba depende de:
    - Tama√±o de la muestra: Muestras m√°s grandes aumentan la potencia.
    - Magnitud del efecto: Efectos m√°s grandes son m√°s f√°ciles de detectar.
    - Nivel de significancia ($\alpha$): Aumentar $\alpha$ incrementa la potencia, pero tambi√©n incrementa el riesgo de Error Tipo I.

- **Mitigaci√≥n:**
  - Asegurarse de que el tama√±o de la muestra sea lo suficientemente grande.
  - Dise√±ar el estudio de manera que los efectos sean medibles con precisi√≥n.

$$
\text{Potencia de la prueba} = 1 - \beta
$$

---

### **Relaci√≥n entre los Errores Tipo I y Tipo II**

Existe una relaci√≥n inversa entre los errores tipo I y tipo II:
- Disminuir $\alpha$ reduce el error de tipo I pero incrementa $\beta$ (error de tipo II).
- Incrementar el tama√±o de la muestra puede disminuir ambos errores, ya que mejora la precisi√≥n de las estimaciones estad√≠sticas.

En la pr√°ctica, el investigador debe equilibrar estos riesgos dependiendo del contexto del problema y las posibles consecuencias de cada tipo de error. Por ejemplo:
- En situaciones cr√≠ticas (e.g., tratamientos m√©dicos), puede ser m√°s importante minimizar el error de tipo I.
- En investigaciones exploratorias, puede ser m√°s importante maximizar la potencia para detectar efectos potenciales.

### <font color='blue'>**Matriz de confusi√≥n**</font>

|   Decisi√≥n basada en la muestra   | $H_0$ es verdadera | $H_0$ es falsa |
|----------------------------------------|:------------------:|:--------------:|
| **No rechazar $H_0$**  | <font color='green'>__Decisi√≥n correcta__</font> <br>(probabilidad = $1-\alpha$)     |<font color='red'>__Error tipo II__ </font> <br> no rechazar $H_0$ cuando es falsa <br>(probabilidad = $\beta$) |
| **Rechazar $H_0$**   | <font color='red'>__Error tipo I__</font>  <br>rechazar $H_0$ cuando es verdadera <br>(probabilidad = $\alpha$)  | <font color='green'>__Decisi√≥n correcta__</font> <br>(probabilidad = $1-\beta$)|


### **Definici√≥n del Efecto**

En estad√≠stica, el **efecto** se refiere a la **diferencia o impacto observado en una variable de inter√©s como resultado de un fen√≥meno o intervenci√≥n espec√≠fica**. El efecto se mide para evaluar si existe una relaci√≥n significativa entre las variables o si los cambios observados son producto del azar.

#### **Definici√≥n formal del efecto**
El efecto puede representarse como la magnitud de la discrepancia entre:
1. **El valor observado en los datos muestrales**, y
2. **El valor esperado bajo la hip√≥tesis nula ($H_0$)**.

El efecto puede cuantificarse en t√©rminos absolutos (como una diferencia de medias) o relativos (como un porcentaje o cociente).

---

### **Tipos de Efecto**
1. **Efecto absoluto:**
   - Ejemplo: Diferencia entre dos medias ($\bar{X_1} - \bar{X_2}$) o proporciones ($\hat{p_1} - \hat{p_2}$).
   - Representa el cambio observado en las unidades originales de la variable.

2. **Efecto relativo:**
   - Ejemplo: Raz√≥n de probabilidades (odds ratio) o riesgo relativo.
   - Expresa la magnitud del efecto como un m√∫ltiplo o porcentaje en relaci√≥n a un valor de referencia.

3. **Efecto est√°ndar:**
   - Ejemplo: Tama√±o del efecto ($d$ de Cohen), que mide la magnitud del efecto en unidades de desviaci√≥n est√°ndar, permitiendo comparar efectos entre diferentes estudios.

---

### **Medici√≥n del Efecto**
El efecto se calcula a partir de los datos muestrales y se interpreta en el contexto de una prueba de hip√≥tesis. Es un componente central para:
- **Pruebas de hip√≥tesis:** Determinar si el efecto es estad√≠sticamente significativo.
- **Potencia de la prueba:** Detectar efectos reales en la poblaci√≥n.
- **Tama√±o de muestra:** Planificar estudios que tengan suficiente sensibilidad para identificar el efecto esperado.

---

### **Ejemplo Pr√°ctico**
- En un estudio cl√≠nico, el efecto podr√≠a ser la diferencia promedio en la presi√≥n arterial entre pacientes tratados con un nuevo medicamento y aquellos que recibieron un placebo.
- Si el medicamento reduce la presi√≥n arterial en promedio 10 mmHg m√°s que el placebo, este ser√≠a el efecto absoluto.

---

### **Importancia del Efecto**
El efecto es esencial no solo para establecer la significancia estad√≠stica (si el efecto es diferente de 0), sino tambi√©n para evaluar la **significancia pr√°ctica** o **importancia cl√≠nica**, que considera si el efecto tiene un impacto relevante en el contexto del estudio. Esto asegura que los resultados no solo sean estad√≠sticamente significativos, sino tambi√©n √∫tiles y aplicables en la pr√°ctica.

```
from scipy.stats import norm

# Par√°metros
alpha = 0.05  # Nivel de significancia (error Tipo I)
beta = 0.2    # Probabilidad de error Tipo II
mu_0 = 0  # Media bajo H0
mu_1 = 1  # Media bajo H1 (efecto)
sigma = 1  # Desviaci√≥n est√°ndar

# Definir los puntos cr√≠ticos
z_critical = norm.ppf(1 - alpha)
x = np.linspace(-3, 4, 1000)

# Distribuciones
pdf_H0 = norm.pdf(x, mu_0, sigma)  # Bajo H0
pdf_H1 = norm.pdf(x, mu_1, sigma)  # Bajo H1

# Graficar
plt.figure(figsize=(12, 6))
plt.ylim(0, max(pdf_H1)*1.2)

# Curvas
plt.plot(x, pdf_H0, label="Distribuci√≥n bajo H0 (media=0)", color="blue")
plt.plot(x, pdf_H1, label="Distribuci√≥n bajo H1 (media=1)", color="red")

# √Årea del error tipo I (alpha)
plt.fill_between(x, 0, pdf_H0, where=(x >= z_critical), color="yellow", alpha=0.5, label="Error Tipo I (alpha)")

# √Årea del error tipo II (beta)
plt.fill_between(x, 0, pdf_H1, where=(x <= z_critical), color="green", alpha=0.5, label="Error Tipo II (beta)")

# L√≠neas cr√≠ticas
plt.axvline(z_critical, color="black", linestyle="--", label=f"Valor cr√≠tico Z = {z_critical:.2f}")

# Indicador del efecto (l√≠nea horizontal con bordes)
pos_y = max(pdf_H1) * 1.05
plt.hlines(y=pos_y, xmin=mu_0, xmax=mu_1, color="purple", linestyle="--", label="Efecto (media bajo H1 - media bajo H0)")
plt.scatter([mu_0, mu_1], [pos_y, pos_y], color="purple", s=50)  # Bordes del efecto

# Etiquetas del efecto
plt.text((mu_0 + mu_1) / 2, pos_y*1.05, "Tama√±o del efecto", color="purple", fontsize=10, ha="center")

# Etiquetas y leyenda
plt.title("Errores Tipo I y Tipo II con Visualizaci√≥n del Efecto", fontsize=14)
plt.xlabel("Valor de la variable", fontsize=12)
plt.ylabel("Densidad", fontsize=12)
plt.grid()

# Posicionar leyenda fuera del gr√°fico
plt.legend(bbox_to_anchor=(1, 1), loc="upper left", fontsize=10)
plt.tight_layout()

plt.show()
```
