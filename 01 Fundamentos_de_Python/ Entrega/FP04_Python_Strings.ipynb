## <font color='blue'>**_Indexing_ y _Slicing_ sobre strings**</font>

NOTA: Intensionalmente no traduciremos conceptos en inglés para habituarnos en su uso y hacer más fácil la comprensión de otros documentos a los que accedamos.

Recuerde que las cadenas son *secuencias ordenadas* de caracteres. Esto significa que podemos seleccionar caracteres individuales (__indexing__) o tomar subsecciones de la cadena (__slicing__).

### Indexing

Indexing comienza a contar desde 0, de esta forma el string **hello**:

    character:    h    e    l   l   o
    index:        0    1    2   3   4
    
Puedes usar corchetes cuadrados para seleccionar un caracter

````python
word = "hello"
````

````python
word[0]
````

````python
word[3]
````

````python
# string index out of range
word[10]
````

Python soporta indexación reversa (reverse indexing):

    character:        h     e     l    l    o
    index:            0     1     2    3    4
    reverse index:    0    -4    -3   -2   -1  
    
Reverse indexing es usado comúnmente para selecccionar caractares desde el final de una secuencia:

````python
word[-1]
````

### Slicing

Podemos tomar subsecciones completas de una cadena con notación *slice*.

Esta es la notación:

     [inicio: parada: paso]

Aspectos clave a tener en cuenta:

1. La dirección del índice de inicio corresponde a donde comenzará su segmento.
2. El índice de parada corresponde a donde subirá el corte. **¡No incluye este carácter de índice!**
3. El tamaño del paso es la cantidad de caracteres que omite a medida que avanza y toma el siguiente.

Veamos algunos ejemplos

````python
alpha = 'abcdef'
````

````python
# Observa como la "d" no es incluida!
alpha[0:3]
````

````python
alpha[0:4]
````

````python
# Desde la posición 2 hasta la 4; esta última no incluída
alpha[2:4]
````

````python
# Desde la posición 2 hasta el final del string
alpha[2:]
````

````python
# Desde el inicio del string hasta la posición 2; esta última no incluída
alpha[:2]
````

````python
# Desde la posición 0 hasta la 6, con paso 2
# Si pongo un decimal en cualquier índice de slicing me lanzará error
alpha[0:6:2]
````

````python
# Cadena alrevés (del final al comienzo yendo hacia atrás (paso -1))
alpha[-1::-1]
# Desde el final hacia la posición 1
alpha[-1:0:-1]
#
alpha[-1:-4:-1]
#
alpha[-4:-1:1]
````

<font color='purple' style='bold' size=5>**Experimento** </font>


Aprovecharemos esta parte de la actividad para realizar un experimento, explorando la versatilidad de Python para lograr el mismo resultado  de variadas formas.

En conjunto propusimos cuatro formas de obtener el resultado solicitado.

Luego, describimos tres formas de hacer benchmarking del tiempo que tomaba la ejecución:
* Utilizando perf_counter de la librería time
* Con la magic word %time
* Y finalmente, con la libreria timeit

Finalmente, recolectamos la información obtenida y generamos unos gráficos para evaluar qué alternativa fue más rápida, para luego discutir las posibles causas.




````python

#Primero, definimos las variables y una función que realice los prints de las cuatro formas definidas para sacar los dígitos solicitados.

import time

#ruts a extraer dígitos

ruts = {1 : '9.621.955-4',
        2 : '12.190.843-K'}


def slicingRut(rut:str, opcion:int):
  """
  Obtiene los últimos cuatro dígitos antes del dígito verificador de un rut
  Args:

    rut:      Un rut sin espacios, con guión, con . como separador de miles.
              No se realizarán validaciones ni try-except (ya nos salimos bastante del scope del ejercicio).

    opcion:   Cuál de las alternativas para extraer lo deseado se utilizará (1 a 4 hasta el momento).
              Tampoco se realizará validaciones ni try-except.

  """


  opciones = {
      1 : rut.replace('.','')[-6:-2],
      2 : rut[-7]+rut[-5:-2],
      3 : rut.replace('.' , '').replace('-', '')[-5:-1],
      4 : rut.replace('.' , '').split('-')[0][-4:]
      }
  return print(opciones[opcion])

#definimos un dict vacío donde registraremos los tiempos para cada opción
tiempos = {
        1: [],
        2: [],
        3: [],
        4: [],
        }
#Ahora un for loop para probar cada forma de medición del tiempo


# Utilizando perf_counter, se ejecutará 1000 veces cada rut con cada método

for i in range(1,len(ruts)+1): #len para adaptarse a la cantidad de ruts en 'ruts'
  for j in range (1,5):
    for k in range(1000):
      print(f"Rut {i}, Opción {j}")

      inicio=time.perf_counter()
      slicingRut(ruts[i], j)
      fin=time.perf_counter()
      total = fin - inicio
      print(f"Esto tomó {total} segundos usando perf_counter()")
      tiempos[j].append(total)

"""
# Utilizando la magic word %time

for i in range(1,len(ruts)+1): #len para adaptarse a la cantidad de ruts en 'ruts'
  for j in range (1,5):
    print(f"Rut {i}, Opción {j}")

    tiempo = %time slicingRut(ruts[i], j)


    print(f"Esto tomó {tiempo} segundos usando %time")

"""
# Utilizando timeit, se ejecutará 1000 veces cada rut con cada método

import timeit

for i in range(1,len(ruts)+1): #len para adaptarse a la cantidad de ruts en 'ruts'
  for j in range (1,5):
    for k in range(1000):
      print(f"Rut {i}, Opción {j}")
      tiempo = timeit.timeit("lambda:slicingRut(ruts[i], j)", number=1)
      print(f"Esto tomó {tiempo} segundos usando timeit.")
````

Obtenemos la cantidad de ejecuciones que se realizó de cada alternativa de resolución para confirmar que fueron 2000

````python
len(tiempos[1])
````

Ahora, graficamos los tiempos de ejecución de cada alternativa de resolución en un boxplot con la librería matplotlib.

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame.from_dict(tiempos)
df.plot(kind='box')

plt.show()
````

A primera vista, se podría concluir que la alternativa de resolución 2 es la que tiene los menores tiempo de ejecución, pero para revisarlo de manera más precisa utilizamos la función describe para obtener los descriptores estadísticos de los tiempos de ejecución.

````python
df.describe()
````

#### Resultados
Podemos observar que rut.replace('.' , '').replace('-', '')[-5:-1] es la forma que tomó menos tiempo al observar que su media fue de 0.00154 segundos.

#### Conclusiones

A pesar de que se logró demostrar que la alternativa de solución 2 entregaba los menores tiempos de ejecución, podemos concluir que con códigos simples como estos no hay una diferencia relevante entre los distintos métodos. Gran parte de las variaciones en los tiempos para estos casos están asociados a los momentos de ejecución de cada iteración, lo que depende de factores del sistema.
