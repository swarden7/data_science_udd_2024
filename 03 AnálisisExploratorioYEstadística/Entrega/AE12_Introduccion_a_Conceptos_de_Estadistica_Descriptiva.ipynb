## <font color='blue'>**Introducción**</font>

### **¿Qué es el Análisis Exploratorio de Datos?**

El __Análisis Exploratorio de Datos__ o (__EDA__) consiste en una serie de técnicas y herramientas que nos permiten comprender los conjuntos de datos resumiendo sus características principales, a menudo graficándolos visualmente y utilizando descriptores estadisticos. Este paso es muy importante especialmente cuando llegamos a modelar los datos para aplicar el aprendizaje automático. La visualización en EDA consta de histogramas, boxplots, diagrama de dispersión y muchos más. A menudo se necesita mucho tiempo para explorar los datos. A través del proceso de EDA, podemos definir el enunciado del problema, desarrollar nuestras preguntas de investigación, realizar el diseño experimental, entre otrsa cosas.

La **Estadística** es una disciplina científica que se ocupa de la obtención, ordenación  y análisis de un conjunto de datos con el fin de obtener explicaciones y predicciones sobre un fenómeno en estudio.



## <font color='blue'>**Conceptos básicos**</font>

**Población** : conjunto sobre el que estamos interesados en obtener conclusiones y  hacer inferencias.

**Muestra** : subconjunto de la población, sobre la cual se harán las observaciones  o mediciones.

**Variable**	: es una característica de un objeto o sujeto, que va variando entre los   diferentes individuos  de una población.

**Dato** : valores que se obtienen al realizar un estudio estadístico.

**Parámetro** : cantidad numérica que describe una característica de una Población.

**Estadístico** : cantidad numérica que describe una característica de una Muestra.

La **inferencia estadística** está basada en el estudio de las **muestras**. La muestra debe ser **representativa de la población** para poder extraer conclusiones válidas sobre la misma. La muestra debe ser **aleatoria**.

## <font color='blue'>**Medidas de Tendencia Central**</font>

Las medidas de tendencia central son  valores que se encuentra hacia el centro de la distribución de los valores de una serie de observaciones.

La **Moda** de una variable Cualitativa o Cuantitativa, es el valor que se presenta con mayor frecuencia en una distribución de datos.

1. Cuando __dos datos__ se presentan con la misma frecuencia y ésta es la más alta, ambos datos son moda, es decir,  la distribución es bimodal.
2. Cuando __dos o más datos__ se presentan con la misma frecuencia y esta es la más alta, todos datos son moda, es decir, la distribución es multimodal.
3. Cuando __todos los datos__ tienen la misma frecuencia, se dice que no hay moda.

La **Mediana** de una variable cualitativa ordinal o cuantitativa, corresponde al valor intermedio, cuando los valores de la variable se presentan en orden de magnitud creciente (o decreciente).


$$ Me = X_{\frac{n}{2}}, \mbox{  Para el caso que n es impar.}  $$

$$ Me = \frac{X_{\frac{n}{2}} + X_{\frac{n}{2}+1}}{2},  \mbox{  Para el caso que n es par.}$$

La **media aritmetica** de una variable cuantitativa, se calcula sumando los valores y dividiéndolos por el total de ellos.

$$ Media = \frac{\sum X_i}{n}$$

**Rango**: No es una medida significativa, pues solo indica cuán dispersos están los datos entre los extremos.

$$R = X_{max} - X_{min}$$

**Coeficiente de variación**: También llamado _Coeficiente de Variación de Pearson_, es una medida estadística que nos informa acerca de la dispersión relativa de un conjunto de datos. Es decir, nos informa, al igual que otras medidas de dispersión, si una variable se mueve mucho, poco, más o menos que otra. Se utiliza cuando se necesita comparar la dispersión de distribuciones de variables que tienen distintas unidades de medida.

$$CV = \frac{S}{\bar{X}}$$

Su fórmula expresa la desviación estándar como porcentaje de la media aritmética, mostrando una interpretación relativa del grado de variabilidad, independiente de la escala de la variable, a diferencia de la desviación típica o estándar. Por otro lado presenta problemas ya que a diferencia de la desviación típica este coeficiente es fuertemente sensible ante cambios de origen en la variable. Por ello es importante que todos los valores sean positivos y su media dé, por tanto, un valor positivo. A mayor valor del coeficiente de variación mayor heterogeneidad de los valores de la variable; y a menor C.V., mayor homogeneidad en los valores de la variable. Por ejemplo, si el C.V es menor o igual al 30%, significa que la media aritmética es representativa del conjunto de datos, por ende el conjunto de datos es "Homogéneo". ## <font color='blue'>**Medidas de posición**</font>

**Cuartiles**  son medidas de posición  que dividen  la distribución ordenada en cuatro partes iguales, ellos son $Q_1 , Q_2 , Q_3 $.

$Q_1$: ***Cuartil 1***. Es el valor de la variable que deja a la izquierda el 25% de los datos.

**Rango intercuartilico (IQR)**: es una estimación estadística de dispersión  de una distribución de datos y corresponde a:  

$$IQR = Q_3-Q_1$$

**Percentiles** son medidas de posición que dividen la distribución ordenada en 100 partes iguales, ellos son $P_1$, $P_2$, ..., $P_k$ , con $k=[1,99]$.

$P_K$ :  ***Percentil k***. Es el valor de la variable que deja a la izquierda el $k\%$ de los datos.

## <font color='blue'>__Asimetría__</font>

La __asimetría__ es una medida de la falta de simetría o sesgo en la distribución de un conjunto de datos. Una distribución es simétrica si se ve igual a ambos lados de su media.

Tipos de Asimetría

- __Asimetría o Sesgo Positivo (Derecha)__: Cuando la cola en el lado derecho de la distribución es más larga o más gruesa que la izquierda.
- __Asimetría o Sesgo Negativo (Izquierda)__: Cuando la cola en el lado izquierdo de la distribución es más larga o más gruesa que la derecha.
- Un valor de cero indica que no hay ningún sesgo en la distribución, lo que significa que la distribución es perfectamente simétrica, y que, por lo tanto, su media y mediana son iguales.

<img src="https://drive.google.com/uc?export=view&id=1sHNxKTjFE32UHmjH4vAennEQjCd3tjE-" >

`````python
from scipy.stats import skew, norm, gamma, beta
import scipy.stats as stats

simetrica = norm.rvs(size=1000)
a = 1.99
derecha = gamma.rvs(a, size=1000)
a, b = 2.31, 0.627
izquierda = beta.rvs(a, b, size=1000)

print(f'Asimetría Positiva: {skew(derecha):.2f}')
print(f'Asimetría Negativa: {skew(izquierda):.2f}')
print(f'Siméttica: {skew(simetrica):.2f}')
`````

`````python
# Gráficos
import seaborn as sns
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
sns.histplot(derecha, bins=50, kde=True)
plt.title(f'Asimetría Positiva: {skew(derecha):.2f}')
plt.subplot(1, 3, 2)
sns.histplot(simetrica, bins=50, kde=True)
plt.title(f'Simétrica: {skew(simetrica):.2f}')
plt.subplot(1, 3, 3)
sns.histplot(izquierda, bins=50, kde=True)
plt.title(f'Asimetría Negativa: {skew(izquierda):.2f}')

plt.tight_layout()
plt.show()
`````


## <font color='blue'>__Curtosis__</font>

La curtosis es una medida de la "agudeza" de una distribución. Indica la cantidad de datos que están cerca de la media en comparación con una distribución normal.

Tipos de Curtosis

- __Leptocúrtica__: Distribución con colas más pesadas y un pico más agudo que la distribución normal. Curtosis > 0 (según definición de Fisher), o Cursosis > 3 (según definición de Pearson).

- __Mesocúrtica__: Distribución con la misma forma que la distribución normal. Curtosis = 0 (según definición de Fisher), o Cursosis = 3 (según definición de Pearson).

- __Platicúrtica__: Distribución con colas más ligeras y un pico más plano que la distribución normal. Curtosis < 0 (según definición de Fisher), o Cursosis < 3 (según definición de Pearson).


<img src="https://drive.google.com/uc?export=view&id=1XbtCQSp7yFjJKmxVLWC2-vCIss0t5jHE" >



`````python
from scipy.stats import kurtosis, norm, uniform, laplace
import scipy.stats as stats

meso_data = norm.rvs(size=1000)
lepto_data = laplace.rvs(size=1000)
plati_data = uniform.rvs(size=1000,loc=-2, scale=4)

# Calcular curtosis
lepto_kurtosis = kurtosis(lepto_data, fisher=True)
plati_kurtosis = kurtosis(plati_data, fisher=True)
meso_kurtosis = kurtosis(meso_data, fisher=True)


print(f'Curtosis Leptocúrtica: {lepto_kurtosis:.2f}')
print(f'Curtosis Platicúrtica: {plati_kurtosis:.2f}')
print(f'Curtosis Mesocúrtica: {meso_kurtosis:.2f}')
`````

`````python
# Veámoslo gráficamente
x = np.linspace(-5, 5, 100)
ax = plt.subplot()
dist_name = ['laplace', 'norm', 'uniform']
for d in dist_name:
    if d == 'uniform':
        dist = getattr(stats, d)(loc=-2, scale=4)
    else:
        dist = getattr(stats, d)
    data = dist.rvs(size=1000)
    kur = kurtosis(data, fisher=True)
    y = dist.pdf(x)
    ax.plot(x, y, label=f"{d}, {kur: 4.2f}")
    ax.legend()
`````

### __Estadística Descriptiva__

La estadística descriptiva se utiliza para describir, resumir y entender las características básicas de un conjunto de datos, sin hacer inferencias a partir de él.

__Conceptos Básicos__

- __Población__
La población se refiere al conjunto completo de elementos o casos que se están estudiando.
- __Muestra__
Una muestra es un subconjunto de la población.
- __Variable__
Una variable es cualquier característica que pueda variar y se pueda medir en una población.
- __Dato__
Un dato es una observación o medida de una variable.

__Medidas de Centralidad__

- __Media__
La media es el promedio de un conjunto de datos.
- __Mediana__
La mediana es el valor que separa un conjunto de datos en dos partes iguales.
- __Moda__
La moda es el valor que aparece con mayor frecuencia en un conjunto de datos.

__Medidas de Dispersión__

- __Rango__
El rango es la diferencia entre el valor máximo y el valor mínimo de un conjunto de datos.
- __Varianza__
La varianza es el promedio de las diferencias al cuadrado entre cada dato y la media.
- __Desviación Estándar__
La desviación estándar es la raíz cuadrada de la varianza.

__Medidas de Posición__

- __Cuartiles__
Los cuartiles dividen un conjunto de datos en cuatro partes iguales.
- __Percentiles__
Los percentiles dividen un conjunto de datos en cien partes iguales.

__Error__

- __Error Estándar__
El error estándar es una medida de la variabilidad de la media muestral.
- __Error Absoluto__
El error absoluto es la diferencia entre el valor observado y el valor verdadero.
- __Error Relativo__
El error relativo es el error absoluto dividido por el valor verdadero.

__Otros Conceptos__

- __Asimetría__
La asimetría es una medida de la falta de simetría en la distribución de un conjunto de datos.
- __Curtosis__
La curtosis es una medida de la "agudeza" de una distribución.
