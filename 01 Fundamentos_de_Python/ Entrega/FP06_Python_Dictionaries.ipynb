Hasta ahora solo hemos visto cómo almacenar tipos de datos en secuencias tales como almacenar caracteres en una cadena (string) o elementos en una lista (list). Pero, ¿y si queremos almacenar información de otra manera? La mayoría de los lenguajes de programación tienen lo que se llama una *tabla hash*, que es una combinación de elementos clave-valor. Bajo el capó, este tipo de estructuras permite tiempos de búsqueda de objetos más rápidos cuando las estructuras tienen muchos elementos.

Un diccionario en Python es una estructura de datos que permite almacenar y organizar pares de clave-valor. Es una colección __mutable__, __indexada__ y __no ordenada__ de elementos. En lugar de utilizar índices numéricos como en las listas, los diccionarios utilizan claves para acceder y almacenar valores asociados.

Las características principales de los diccionarios en Python son:

* Clave única: Cada clave en un diccionario es única y se utiliza para acceder al valor asociado.
* Mutable: Los diccionarios pueden modificarse agregando, modificando o eliminando pares clave-valor.
* No ordenado: A diferencia de las listas, los diccionarios no mantienen un orden específico de los elementos.
* Colección de pares clave-valor: Cada elemento del diccionario está compuesto por una clave y su correspondiente valor.
* Puede contener diferentes tipos de datos: Tanto las claves como los valores pueden ser de cualquier tipo de dato válido en Python, como enteros, cadenas, listas u otros diccionarios.

## <font color='PURPLE'>__EXPERIMENTO__:</font>

 ### Explorando alternativas de unión de diccionarios

En el siguiente ejemplo, podemos confirmar la utilización de operadores matemáticos como la adición (`+`)para la unión de 2 listas:

````python
list_a = [1, 2, 3, 4, 5]
list_b = [3, 4, 5, 6, 7, 8]
print (list_a + list_b)
````

Sin embargo, basta con seguir un paso más en la exploración para encontrarnos con que no es posible hacer sustracciones de listas.

````python
#El siguiente código arrojará un ERROR
print(list_a - list_b)
````

Entonces, podemos inferir que podríamos hacer una adición de 2 diccionarios, siguiendo la misma dinámica de las listas, utilizando el operador `+` correcto?
Hagamos la prueba:

````python
#El siguiente código arrojará un ERROR
dict_a = {'a': 1, 'b': 2, 'c': 3}
dict_b = {'d': 4, 'e': 5, 'f': 6}
print(dict_a + dict_b)
````

Así como la lista no tiene el operador `-` disponible, los diccionarios no tienen el operador `+` disponible.
En la actividad, usamos el método update() para concatenar diccionarios. Pero tambien es posible usar el operador `|` para **unir** diccionarios.
Hagamos la prueba:

````python
full_dict = dict_a | dict_b
print(full_dict)
````

Otra opción para unir los diccionarios sería usando la asignación directa en un bucle *for*.
Hagamos la prueba:

````python
full_dict = {}
for key, value in dict_a.items():
    full_dict[key] = value
for key, value in dict_b.items():
    full_dict[key] = value
print(full_dict)
````

Si refactorizamos el bucle for, podríamos eliminar la redundancia en la asignación. Quedaría así:

````python
full_dict = {}
for d in (dict_a, dict_b):
    for key, value in d.items():
        full_dict[key] = value
print(full_dict)
````

## <font color='PURPLE'>__MATERIAL ADICIONAL__</font>



Los diccionarios en Python son estructuras de datos que permiten almacenar pares de claves y valores, facilitando el acceso rápido a datos mediante una clave única. Su flexibilidad y capacidad para organizar datos de manera estructurada los convierte en una herramienta fundamental en diversos escenarios de programación. Por ejemplo, los diccionarios son especialmente útiles al trabajar con datos tabulares, ya que permiten organizar la información en un formato similar al de una tabla, donde cada clave del diccionario puede representar una columna y cada valor asociado una lista de datos para esa columna.

Para obtener más información sobre los diccionarios y su uso en Python, la documentación oficial de Python proporciona una explicación detallada de sus funcionalidades y algunos ejemplos prácticos que pueden ser de gran ayuda para comprender su versatilidad.

https://docs.python.org/3/tutorial/datastructures.html#dictionaries

Más abajo se aprecia un código de ejemplo

````python
from tabulate import tabulate

# Ejemplo de diccionario
data = {
    "Nombre": ["Juan", "Ana", "Luis"],
    "Edad": [23, 21, 30],
    "Ciudad": ["Santiago", "Valparaíso", "Concepción"]
}

# Convertir el diccionario en una lista de filas para la tabla
headers = data.keys()
rows = zip(*data.values())

# Imprimir la tabla
print(tabulate(rows, headers=headers, tablefmt="grid"))
````
