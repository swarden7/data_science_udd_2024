**1. Mapa de calor (Heatmap):**

Identificación de correlaciones entre asistencia, calificaciones y tasas de aprobación.




````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulación de datos
data = pd.DataFrame({
    'Asistencia': np.random.uniform(50, 100, 100),
    'Promedio_Notas': np.random.uniform(3.0, 7.0, 100),
    'Tasa_Aprobacion': np.random.uniform(60, 100, 100)
})

# Calcular correlaciones
correlation_matrix = data.corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlación entre indicadores académicos")
plt.show()
````

**Resultado:** Un mapa de calor que muestra cómo las variables están relacionadas. Por ejemplo, podrías descubrir que mayor asistencia tiene una fuerte correlación con tasas de aprobación.

El programa CERO es un programa orientado a la prevención de caries en niños de 0 a 9 años, que busca realizar controles preventivos en niños de esas edades, determinando el riesgo cariogénico y según aquello determinando la fecha aproximada del próximo control, el cual puede ser 6 o 12 meses después.

from datetime import date, time

def meses_transcurridos(born):
    today = date.today()
    return (today.year - born.year) * 12 + today.month - born.month - (today.day < born.day)

bajocontrol_df['MESES ULTIMO CONTROL'] = bajocontrol_df['FECHA ESTIMADA PROXIMO CONTROL'].apply(meses_transcurridos)
