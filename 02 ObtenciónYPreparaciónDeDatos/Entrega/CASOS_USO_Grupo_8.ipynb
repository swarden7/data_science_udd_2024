## ***Análisis de la Capacidad de Retención de Oxígeno en la Aireación de Reactores Biológicos en PTARs***

En una Planta de Tratamiento de Aguas Residuales (PTAR), la aireación de los reactores biológicos es fundamental para suministrar oxígeno disuelto (OD) a los microorganismos que descomponen la materia orgánica en el agua. Un porcentaje importante de los sólidos suspendidos totales (SST) en los reactores representa la biomasa activa en el licor mezcla y afecta directamente el consumo de OD. Un aumento en SST indica más microorganismos, lo que generalmente incrementa el consumo de oxígeno y puede reducir el OD disponible.

La eficiencia de transferencia de oxígeno también depende de los niveles de SST y la temperatura: concentraciones altas de SST vuelven el licor mezcla más denso, dificultando la disolución del oxígeno, mientras que temperaturas elevadas disminuyen la solubilidad del oxígeno y aumentan la demanda metabólica.

**Objetivo del Estudio**

Este estudio busca analizar cómo los SST y la temperatura influyen en la "capacidad de retención de oxígeno", usando como indicador la relación OD/flujo de aire. Si el flujo de aire se mantiene constante, un OD bajo indicaría mayor consumo o menor eficiencia en la retención de oxígeno, mientras que un OD alto sugeriría condiciones más favorables para la disolución del oxígeno.

**Estructura del Estudio de Datos**

Los datos a analizar provienen de distintas fuentes con diferentes frecuencias de muestreo:

* Flujo de aire y temperatura: Medidos en línea mediante sensores, y disponibles en archivos CSV.
* Oxígeno disuelto (OD) y SST: Medidos manualmente tres veces al día (mañana, tarde, noche) y registrados en archivos Excel.

**Enfoque del Análisis**

Para este análisis, se combinarán y limpiarán los datos de las diferentes fuentes para obtener una base unificada. Se calculará un indicador de retención de oxígeno para cada turno, definido como la relación entre OD y flujo de aire, y se evaluará su relación con los parámetros de SST y temperatura.

**Estructura del Pseudocódigo para el Análisis de Datos**

1. Descargar y cargar los datos:

Leer los archivos CSV (flujo de aire y temperatura) y Excel (OD y SST) en dataframes.

2. Eliminar outliers en datos de sensores:

En los datos de flujo de aire y temperatura, eliminar valores fuera del percentil 99 y por debajo del percentil 1.

3. Calcular promedios y totales por turno:

Calcular el flujo de aire total y la temperatura promedio entre turnos (mañana, tarde, noche).
Asignar el valor resultante al turno correspondiente (por ejemplo, entre 09:00 y 15:00 → asignar al turno de las 15:00).

4. Procesar datos de OD y SST:

Asegurarse de que el índice de tiempo en el dataframe de OD y SST corresponde al turno en el que se tomó la medición (aproximando la hora a los valores más cercanos: 09:00, 15:00, 21:00).
Definir el índice como una combinación de la fecha y la hora del turno.

5. Unir los dataframes:

Concatenar los dataframes a lo largo de axis=1 para unir todos los datos en una sola tabla por turno.

6. Calcular el indicador de retención

Definido por el oxígeno disuelto dividido en el total de flujo de aire administrado en el turno.

### <font color='blue'> 1. Código de la Solución

````python
## Importamos las bibliotecas necesarias
import pandas as pd
import numpy as np

# 1. Cargar los archivos CSV y Excel
flujo_temp_df = pd.read_csv("nombre_archivo_csv.csv")     # Archivo de flujo y temperatura
od_sst_df = pd.read_excel("nombre_archivo_excel.xlsx")     # Archivo de OD y SST

# 2. Preprocesamiento de los datos de sensores
def eliminar_outliers(df, cols):
    for col in cols:
        q1 = df[col].quantile(0.01)
        q99 = df[col].quantile(0.99)
        df = df[(df[col] >= q1) & (df[col] <= q99)]
    return df

flujo_temp_df = eliminar_outliers(flujo_temp_df, ["Flujo de Aire", "Temperatura"])

# 3. Calcular valores representativos entre turnos
flujo_temp_df["Flujo Turno"] = flujo_temp_df["Flujo de Aire"].resample("6H").sum()
flujo_temp_df["Temp Promedio Turno"] = flujo_temp_df["Temperatura"].resample("6H").mean()
# Nota: resample funciona cuando el index es de tipo Datetime

# 4. Procesamiento de los datos de OD y SST
# Suponiendo que 'Fecha' y 'Hora' son columnas en od_sst_df
od_sst_df["Fecha-Hora"] = pd.to_datetime(od_sst_df["Fecha"].astype(str) + " " + od_sst_df["Hora"].astype(str))
od_sst_df.set_index("Fecha-Hora", inplace=True)

# Aproximar horas a los turnos más cercanos
od_sst_df.index = od_sst_df.index.round("6H")

# Paso 5: Concatenación de los datos
# Concatenar ambos DataFrames en el eje de las columnas
datos_concatenados = pd.concat([flujo_temp_df, od_sst_df], axis=1)

# Paso 5: Calcular el indicador de retención
datos_concatenados["Indicador de Retención"] = datos_concatenados["OD"] / datos_concatenados["Flujo Turno"]
````

Este planteamiento permite organizar los datos de manera que se pueden realizar análisis exploratorios y modelado para comprender la influencia de diferentes factores en la aireación de los reactores biológicos de la PTAR.

Como paso inicial en el análisis, se podría generar una matriz de correlación para explorar la relación entre OD/flujo de aire, SST y temperatura, y así entender cómo estos factores afectan la capacidad de retención de oxígeno en el licor mezcla.

Este tipo de análisis en las PTARs es importante, ya que se ha observado que en temporadas de invierno, al mantener el mismo protocolo de aireación que en verano, los valores de OD en los reactores aumentan significativamente. Esto se debe a que el agua de entrada presenta una menor carga orgánica y las temperaturas son más bajas, lo que facilita la disolución de oxígeno. Tener un OD más alto de lo necesario en los reactores implica un consumo energético innecesario en la aireación. Dado que la aireación representa entre el 40% y el 50% del consumo energético total en las PTAR, este es un punto clave para la optimización de procesos.

### **Segunda etapa (Desarrollo de código en Python)**

Lectura y Preparación de Datos




````python
import pandas as pd
import numpy as np

def cargar_datos(filepath):
    try:
        datos = pd.read_csv(filepath)
        print("Datos cargados con éxito.")
        return datos
    except FileNotFoundError:
        print("Error: Archivo no encontrado.")
    except Exception as e:
        print(f"Error al cargar los datos: {e}")
    return None

def preparar_datos(datos):
    # Revisión de valores nulos
    datos.fillna({'Asistencia': 0, 'Nota': 0}, inplace=True)

    # Reindexación
    datos.reset_index(drop=True, inplace=True)

    # Normalización de asistencia
    datos['Asistencia'] = datos['Asistencia'] / 100.0
    return datos
````

**Análisis Exploratorio**

````python
def analizar_datos(datos):
    # Agregar métricas clave
    promedio_por_carrera = datos.groupby('Carrera')['Nota'].mean()
    asistencia_promedio = datos['Asistencia'].mean()

    print("Promedio por Carrera:")
    print(promedio_por_carrera)
    print(f"Asistencia promedio: {asistencia_promedio * 100:.2f}%")
````

**Identificación de Riesgo de Deserción**

````python
def identificar_riesgo_desercion(datos):
    datos['En_Riesgo'] = np.where((datos['Nota'] < 4) & (datos['Asistencia'] < 0.7), True, False)
    en_riesgo = datos[datos['En_Riesgo']]

    if not en_riesgo.empty:
        print("Estudiantes en riesgo de deserción:")
        print(en_riesgo[['Estudiante', 'Carrera', 'Nota', 'Asistencia']])
    else:
        print("No hay estudiantes en riesgo de deserción.")
````

**Visualización de Datos**

````python
import matplotlib.pyplot as plt

def generar_reporte(datos):
    # Distribución de notas
    plt.figure(figsize=(10, 6))
    datos['Nota'].plot(kind='hist', bins=10, alpha=0.7, color='blue')
    plt.title("Distribución de Notas")
    plt.xlabel("Nota")
    plt.ylabel("Frecuencia")
    plt.grid(axis='y')
    plt.show()
````

**Ejecución del Flujo Completo**

````python
# Ruta del archivo de datos
ruta = "rendimiento_academico.csv"

# Cargar datos
datos = cargar_datos(ruta)

if datos is not None:
    # Preparar datos
    datos = preparar_datos(datos)

    # Análisis exploratorio
    analizar_datos(datos)

    # Identificación de riesgo
    identificar_riesgo_desercion(datos)

    # Generar visualizaciones
    generar_reporte(datos)
````
