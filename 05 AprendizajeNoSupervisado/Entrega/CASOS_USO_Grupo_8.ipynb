## *** Identificación de Días con Condiciones Similares en Reactores Biológicos de una PTAR***

En el caso de uso anterior, se evaluó el impacto del set-point de oxígeno disuelto (OD) en el consumo energético diario de los reactores biológicos de una Planta de Tratamiento de Aguas Residuales (PTAR). Se aplicaron métodos de inferencia estadística para determinar si la modificación del set-point de OD generaba una reducción significativa en el consumo energético.

Ahora, en esta siguiente etapa del análisis, buscamos profundizar en los patrones de consumo energético utilizando técnicas de aprendizaje no supervisado. Estas técnicas permitirán identificar grupos de días operativos con características similares y determinar si hay patrones que favorecen una mayor eficiencia. Esto facilitará la optimización de la estrategia de aireación y la reducción del consumo energético sin comprometer el rendimiento del tratamiento biológico.

**Objetivo:**

Agrupar días operativos según similitud en sus condiciones de operación, considerando variables como set-point de OD, temperatura, sólidos suspendidos totales (SST), tiempo de aireación y caudal tratado.


**Datos disponibles:**

Disponemos de datos históricos del consumo energético diario (variable de interés) de los reactores biológicos operando bajo determinados set-points de oxígeno disuelto (OD). Además, se cuenta con datos de temperatura del agua, concentración de sólidos suspendidos totales (SST), tiempo de prendido de los aireadores, caudal tratado, entre otras.


**Explicación de los pasos del código:**

1. Preparación de los datos: Procesar los datos históricos y del periodo de prueba. Identificar valores atípicos o inconsistencias en los datos de consumo energético y corregir.

2. Análisis exploratorio: Calcular estadísticas descriptivas (media, desviación estándar, etc.).

3. Aplicación de Clustering para Identificación de Días Similares:

   - K-means.

   - DBSCAN.


4. Evaluar clusterización con métricas cuantitativas.

5. Reducción de Dimensionalidad con PCA: Reduce las variables a 2 dimensiones para visualizar mejor los clusters.

6. . Visualización de Resultados: Gráficos para entender cómo se agrupan los días según condiciones operacionales.

7.. Análisis de los Resultados: Evaluar si ciertos grupos de días están asociados con menor consumo energético. Identificar patrones en los días de menor consumo.

7. Conclusiones y Recomendaciones: Sugerir ajustes operacionales basados en los clusters obtenidos.

### <font color='blue'> 1. Pseudocódigo de la Solución

````python
# 1. Importar bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, homogeneity_score, completeness_score

# 2. Cargar los datos
def cargar_datos(ruta_archivo):
    datos = pd.read_csv(ruta_archivo)
    return datos

datos = cargar_datos("archivo_datos.csv")

# 3. Preprocesamiento de Datos
def limpiar_y_normalizar(datos):
    # Eliminar valores nulos
    datos = datos.dropna()

    # Eliminar valores atípicos usando el rango intercuartil (IQR)
    Q1 = datos.quantile(0.25)
    Q3 = datos.quantile(0.75)
    IQR = Q3 - Q1
    datos = datos[~((datos < (Q1 - 1.5 * IQR)) | (datos > (Q3 + 1.5 * IQR))).any(axis=1)]

    # Normalizar las variables
    scaler = StandardScaler()
    datos_normalizados = scaler.fit_transform(datos)

    return pd.DataFrame(datos_normalizados, columns=datos.columns)

datos = limpiar_y_normalizar(datos)

# 4. Aplicar Clustering para Agrupar Días Similares

# a) K-means
def aplicar_kmeans(datos, n_clusters=3):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(datos)
    return clusters, kmeans

clusters_kmeans, modelo_kmeans = aplicar_kmeans(datos)

# b) DBSCAN
def aplicar_dbscan(datos, eps=0.5, min_samples=5):
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    clusters = dbscan.fit_predict(datos)
    return clusters, dbscan

clusters_dbscan, modelo_dbscan = aplicar_dbscan(datos)

# 5. Reducción de Dimensionalidad con PCA para Visualización
def aplicar_pca(datos, n_componentes=2):
    pca = PCA(n_components=n_componentes)
    datos_pca = pca.fit_transform(datos)
    return datos_pca

datos_pca = aplicar_pca(datos)

# 6. Visualización de Clusters
def visualizar_clusters(datos_pca, clusters, titulo="Clusters Identificados"):
    plt.figure(figsize=(8, 6))
    plt.scatter(datos_pca[:, 0], datos_pca[:, 1], c=clusters, cmap='viridis', alpha=0.7)
    plt.xlabel("Componente Principal 1")
    plt.ylabel("Componente Principal 2")
    plt.title(titulo)
    plt.colorbar()
    plt.show()

visualizar_clusters(datos_pca, clusters_kmeans, "Clusters K-means")
visualizar_clusters(datos_pca, clusters_dbscan, "Clusters DBSCAN")

# 7. Evaluación de los Métodos de Clustering
def evaluar_clusterizacion(datos, clusters, metodo):
    # Silhouette Score (Mide qué tan bien se agrupan los puntos en cada cluster)
    if len(set(clusters)) > 1:  # Silhouette no se puede calcular si hay solo un cluster
        silhouette = silhouette_score(datos, clusters)
    else:
        silhouette = None

    # Homogeneity Score (Mide si todos los clusters contienen solo un tipo de datos)
    homogeneity = homogeneity_score(clusters, clusters)  # No hay etiquetas reales, comparamos consigo mismo

    # Completeness Score (Mide si todos los datos de un grupo están en el mismo cluster)
    completeness = completeness_score(clusters, clusters)  # Similar a homogeneity en este caso

    print(f"Evaluación de {metodo}:")
    print(f" - Silhouette Score: {silhouette if silhouette else 'No aplicable'}")
    print(f" - Homogeneity Score: {homogeneity:.4f}")
    print(f" - Completeness Score: {completeness:.4f}")
    print("------------------------------------------------")

evaluar_clusterizacion(datos, clusters_kmeans, "K-means")
evaluar_clusterizacion(datos, clusters_dbscan, "DBSCAN")

# 8. Análisis de los Clusters
def analizar_clusters(clusters, datos):
    datos["Cluster"] = clusters
    resumen_clusters = datos.groupby("Cluster").mean()
    print("Resumen de Clusters:\n", resumen_clusters)

analizar_clusters(clusters_kmeans, datos)

# 9. Conclusiones y Recomendaciones
def generar_reporte():
    print("Conclusiones:")
    print("- Se identificaron días con condiciones similares en la PTAR.")
    print("- Se detectaron posibles anomalías en el consumo energético.")
    print("- Se pueden ajustar estrategias operacionales basadas en los grupos obtenidos.")

generar_reporte()
````

Este análisis determina la segmentación de días con condiciones similares, permitiendo analizar cómo influyen en el consumo energético.También, permite identificar días atípicos en los que el consumo energético se desvía del comportamiento general.

Estos resultados son base para futuras optimizaciones en la estrategia de aireación, especialmente en el desarrollo de modelos de IA para la optimización en base a la identificación de patrones de oepración.

**Paso 1: Preparación de los Datos**

Supongamos que contamos con un DataFrame df que contiene variables clave (por ejemplo, calificacion, asistencia, participacion).





````python
import pandas as pd
import numpy as np

# Ejemplo de datos simulados
np.random.seed(42)
df = pd.DataFrame({
    'calificacion': np.random.normal(5, 1, 200),
    'asistencia': np.random.uniform(70, 100, 200),
    'participacion': np.random.uniform(1, 10, 200)
})

# Visualización de las primeras filas
print(df.head())
````

**Paso 2: Aplicación de PCA para Reducción de Dimensionalidad**

*Utilizaremos PCA para transformar las variables originales en componentes principales, facilitando la visualización y comprensión de la estructura subyacente.

````python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Estandarización de los datos (recomendado para PCA)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Aplicar PCA para reducir a 2 componentes
pca = PCA(n_components=2)
pca_components = pca.fit_transform(df_scaled)

# Crear un DataFrame con los componentes principales
df_pca = pd.DataFrame(data=pca_components, columns=['PC1', 'PC2'])
print(df_pca.head())

# Visualización de la varianza explicada
print("Varianza explicada por cada componente:", pca.explained_variance_ratio_)

# Gráfico de dispersión de los componentes
plt.figure(figsize=(8,6))
plt.scatter(df_pca['PC1'], df_pca['PC2'], alpha=0.7)
plt.title('Visualización de Componentes Principales')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
````

**Paso 3: Segmentación con K-Means**

Con los componentes principales, aplicaremos el algoritmo K-Means para segmentar a los estudiantes en grupos homogéneos.

````python
from sklearn.cluster import KMeans

# Definir el número de clusters (por ejemplo, 3)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(df_pca)

# Añadir la información de clusters al DataFrame PCA
df_pca['Cluster'] = clusters

# Visualización de los clusters
plt.figure(figsize=(8,6))
plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df_pca['Cluster'], cmap='viridis', alpha=0.7)
plt.title('Segmentación de Estudiantes con K-Means')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar(label='Cluster')
plt.show()

# Análisis de clusters: estadísticas descriptivas por grupo
df['Cluster'] = clusters
print(df.groupby('Cluster').mean())
````

**Conclusiones y Aplicaciones**

**Visualización y reducción de la complejidad:** PCA nos permite observar la estructura interna de los datos y entender cuáles son las variables más influyentes.

**Segmentación de la población estudiantil:** K-Means agrupa a los estudiantes en clusters, lo que puede ayudar a identificar subgrupos con necesidades o características específicas.

**Implementación estratégica:** Los insights derivados de este análisis pueden servir para orientar políticas educativas, personalizar intervenciones y optimizar el uso de recursos.

````python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Separar los grupos
daño_migrantes = df[df['MIGRANTES'] == 'SI']['DAÑO POR CARIES']
daño_no_migrantes = df[df['MIGRANTES'] == 'NO']['DAÑO POR CARIES']

# Estadísticas descriptivas
print("Estadísticas descriptivas:")
print("\nMigrantes:")
print(daño_migrantes.describe())
print("\nNo Migrantes:")
print(daño_no_migrantes.describe())

# Visualización
plt.figure(figsize=(12, 5))

# Boxplot
plt.subplot(1, 2, 1)
sns.boxplot(x='MIGRANTES', y='DAÑO POR CARIES', data=df)
plt.title('Boxplot de Daño por Caries por Estado Migratorio')

# Violin plot
plt.subplot(1, 2, 2)
sns.violinplot(x='MIGRANTES', y='DAÑO POR CARIES', data=df)
plt.title('Violin Plot de Daño por Caries por Estado Migratorio')
plt.tight_layout()
plt.show()

# Test de normalidad
_, p_norm_migrantes = stats.normaltest(daño_migrantes)
_, p_norm_no_migrantes = stats.normaltest(daño_no_migrantes)

print("\nPruebas de normalidad (p-valores):")
print(f"Migrantes: {p_norm_migrantes:.4f}")
if p_norm_migrantes < 0.05:
    print("El daño por caries de los migrantes no sigue una distribución normal")
else:
    print("El daño por caries de los migrantes sigue una distribución normal")

print(f"No Migrantes: {p_norm_no_migrantes:.4f}")
if p_norm_no_migrantes < 0.05:
    print("El daño por caries de los no migrantes no sigue una distribución normal")
else:
    print("El daño por caries de los no migrantes sigue una distribución normal")

print("Observación: Me parece curioso que uno tenga distribución normal y el otro no.\nCreo que puede deberse a lo reducido de la muestra de migrantes (n).\nIndagaría más en qué tan rigurosa es la toma de este valor.")



# Intervalos de confianza (95%)
media_migrantes = np.mean(daño_migrantes)
error_std_migrantes = stats.sem(daño_migrantes)
ic_migrantes = stats.t.interval(0.95, len(daño_migrantes)-1,
                               loc=media_migrantes,
                               scale=error_std_migrantes)

# Para no migrantes
media_no_migrantes = np.mean(daño_no_migrantes)
error_std_no_migrantes = stats.sem(daño_no_migrantes)
ic_no_migrantes = stats.t.interval(0.95, len(daño_no_migrantes)-1,
                                  loc=media_no_migrantes,
                                  scale=error_std_no_migrantes)

print("\nIntervalos de confianza 95%:")
print(f"Migrantes: [{ic_migrantes[0]:.2f}, {ic_migrantes[1]:.2f}]")
print(f"No Migrantes: [{ic_no_migrantes[0]:.2f}, {ic_no_migrantes[1]:.2f}]")

# Tamaño del efecto (Cohen's d)
n1, n2 = len(daño_migrantes), len(daño_no_migrantes)
var1, var2 = np.var(daño_migrantes, ddof=1), np.var(daño_no_migrantes, ddof=1)
pooled_se = np.sqrt(((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2))
cohens_d = (np.mean(daño_migrantes) - np.mean(daño_no_migrantes)) / pooled_se

# Gradación del Cohen's
if cohens_d < 0.2:
    tamaño = "pequeño"
elif 0.2 <= cohens_d < 0.5:
    tamaño = "medio"
else:
    tamaño = "grande"

print(f"\nTamaño del efecto (Cohen's d, diferencia entre ambos grupos): {cohens_d:.4f} ({tamaño})")

print('Observación: Arroja una advertencia por el tamaño de la muestra')
````
