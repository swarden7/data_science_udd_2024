##<font color='blue'>**Variables aleatorias**</font>
Una __variable aleatoria__ es una función que asigna un valor numérico, al resultado de un experimento aleatorio. Una variable aleatoria puede ser __discreta__ o __continua__. Las __variables aleatorias discretas__ son aquellas que presentan un número contable de valores; por ejemplo, el número de personas que viven en una casa (pueden ser 3, 5 o 9). Las __variables aleatorias continuas__ son aquellas que presentan un número incontable de valores; por ejemplo, el peso de las vacas en una granja (una vaca puede pesar 632,12 kg, otra puede pesar 583,12312 kg, otra 253,12012 kg, otra 198,0876 kg y nunca terminaríamos de enumerar todos los posibles valores).

### <font color='blue'>**¿Por qué es importante conocer las distribuciones?**</font>

Muchos resultados en las ciencias se basan en conclusiones que se extraen sobre una población general a partir del estudio de una [**muestra**](https://es.wikipedia.org/wiki/Muestra_estad%C3%ADstica) de esta población. Este proceso se conoce como [**inferencia estadística**](https://es.wikipedia.org/wiki/Estad%C3%ADstica_inferencial); y este tipo de *inferencia* con frecuencia se basa en hacer suposiciones acerca de la forma en que los datos se distribuyen, o requiere realizar alguna transformación de los datos para que se ajusten mejor a alguna de las  [distribuciones](https://es.wikipedia.org/wiki/Distribuci%C3%B3n_de_probabilidad) conocidas y estudiadas en profundidad.

Las distribuciones de probabilidad teóricas son útiles en la inferencia estadística porque sus propiedades y características son conocidas. Si la distribución real de un conjunto de datos dado es razonablemente cercana a la de una distribución de probabilidad teórica, muchos de los cálculos se pueden realizar en los datos reales utilizando hipótesis extraídas de la distribución teórica.

###__Función de Distribución de Probabilidad__
Es un modelo teórico que describe la forma en que varían los resultados de un experimento aleatorio, es decir, nos da todas las probabilidades de todos los posibles resultados que podrían obtenerse cuando se realiza el experimento aleatorio.

Se define como

$$f(x) = P(X = x)$$

Propiedades de Distribución de Probabilidad de una v.a. discreta

$$P(X = x) \gt 0$$

$$ \sum_{i=1}^{n} P(X = x) = 1$$

Nota: Si alguna de estas dos proposiciones no se cumple, la variable NO es una función de probabilidad.

###__Medida de centro y dispersión para una variable aleatoria discreta__

Media probabilística

$$\mu_x = E(X) = \sum_{i=1}^n x_i * p_i $$

Dispersión probabilística

$$\sigma_x^2 = V(X) = \sum_{i=1}^n x_i^2 \; p_i - \mu^2$$

### Explicación Dispersión probabilística ($\sigma_x^2\$)

La dispersión probabilística, o varianza probabilística, mide cuánta variabilidad tienen los valores posibles de una variable aleatoria respecto a su media esperada ($\mu_x$).

#### Derivación:
1. **Definición inicial**: La varianza se define como el valor esperado del cuadrado de las desviaciones respecto a la media:
   $$
   V(X) = E\left[(X - \mu_x)^2\right]
   $$

2. **Expansión del término $(X - \mu_x)^2$**:
   $$
   (X - \mu_x)^2 = X^2 - 2X\mu_x + \mu_x^2
   $$

3. **Sustitución en el valor esperado**:
   $$
   V(X) = \sum_{i=1}^n x_i^2 \cdot p_i - 2\mu_x \sum_{i=1}^n x_i \cdot p_i + \mu_x^2 \sum_{i=1}^n p_i
   $$

4. **Simplificación**:
   - $\sum_{i=1}^n x_i \cdot p_i = \mu_x$
   - $\sum_{i=1}^n p_i = 1$

   Esto reduce la fórmula a:
   $$
   \sigma_x^2 = \sum_{i=1}^n x_i^2 \cdot p_i - \mu_x^2
   $$

#### Interpretación:
- **$\sum_{i=1}^n x_i^2 \cdot p_i$**: Es el valor esperado de los cuadrados de los valores posibles ($x_i^2$).
- **$\mu_x^2$**: Representa el cuadrado de la media probabilística ($\mu_x$).

La varianza mide qué tan dispersos están los valores posibles de $X$ alrededor de su media, ponderados por sus probabilidades.

### <font color='blue'>**Función de Densidad de Probabilidad**</font>

Sea X una variable aleatoria continua y f(x) su Función de Densidad de Probabilidad para dos números $a$ y $b$, con $a \le b$,

$$P(a \le x \le b) = \int_{a}^{b} f(x)dx$$

es decir, la probabilidad de que $X$ asuma un valor en el intervalo $[a,b]$ corresponde al área bajo la curva entre $a$ y $b$.

<img alt="FDP continua" title="FDP continua" src="https://drive.google.com/uc?export=view&id=1XnzwlXZhtWAfOKvEQSvoWvPVnoWMAMMP" high=650px width=650px>

Propiedades de Densidad de Probabilidad de una v.a. continua

$$f(x) \ge 0$$

$$ \int_{-\infty}^{\infty} f(x)dx = 1$$


<br><br>



### <font color='blue'>**Función de Distribución de Probabilidad Acumulada**</font>

Sea X una variable aleatoria continua con función de densidad $f(x)$, se define la función de distribución $F(x)$, como:

$$F(x) = P(X \le x) = \int_{-\infty}^{x} f(x)dx$$

A esta función la llamaremos __función de distribución acumulada__ de la v.a. $x$. Con las siguientes propiedades:

$$F(x) \ge 0$$

$$ \int_{-\infty}^{\infty} f(x)dx = 1$$

<br><br>
Es decir,


$$P(a \le x \le b) = \int_{a}^{b} f(x)dx = F(b) - F(a)$$

### <font color='blue'>**Medida de centro y dispersión para una variale aleatoria continua**</font>

Media probabilística

$$\mu_x = E(X) =\int_{-\infty}^{\infty} x\;f(x)dx$$


Dispersión probabilística (varianza)

$$\sigma_x^2 = V(X) = \int_{-\infty}^{\infty} (x - \mu)^2\;f(x)dx$$


# <font color='purple'>__EXPERIMENTO__</font>

**Simulación de una variable aleatoria continua con distribución normal**

El objetivo será generar un conjunto de datos simulados con una distribución normal, calcular su media y varianza, y graficar la distribución.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simular 1000 datos con distribución normal (media = 0, desviación estándar = 1)
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Calcular media y varianza
mean_data = np.mean(data)
variance_data = np.var(data)

# Imprimir los resultados
print(f"Media de los datos: {mean_data}")
print(f"Varianza de los datos: {variance_data}")

# Graficar la distribución de los datos simulados
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Graficar la curva de la distribución normal teórica
from scipy.stats import norm
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_data, np.sqrt(variance_data))
plt.plot(x, p, 'k', linewidth=2)

plt.title('Distribución normal simulada')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

**Explicación:**

**Simulación de datos:**
Este código genera 1000 datos distribuidos normalmente con una media de 0 y desviación estándar de 1.

**Cálculos:** Se calcula la media y la varianza de los datos generados.

**Gráfica:** Se grafica el histograma de los datos simulados y se sobrepone la curva de la distribución normal teórica basada en la media y varianza calculadas.


Entonces, al superponerse ambas curvas podemos comprobar de manera práctica el teorema del límite central.

**Aplicaciones de la distribución normal**

La distribución normal es utilizada en numerosos campos como la estadística, economía, ingeniería y las ciencias sociales. Algunos ejemplos comunes de aplicaciones incluyen:

**Mediciones en la ingeniería:**

En la producción de piezas de maquinaria, las variaciones en el tamaño de las piezas suelen seguir una distribución normal.

**Evaluación de rendimientos académicos:** Las calificaciones de los estudiantes, si no están sesgadas por factores externos, a menudo siguen una distribución normal, lo que facilita el análisis de los resultados.

**Investigación médica:** En estudios de salud, características como la presión arterial o los niveles de colesterol suelen ser distribuidos normalmente, lo que permite realizar inferencias sobre las poblaciones en estudio.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulamos las calificaciones de 1000 estudiantes (media = 75, desviación estándar = 10)
np.random.seed(42)
calificaciones = np.random.normal(loc=75, scale=10, size=1000)

# Calcular la media y la desviación estándar
media_calificaciones = np.mean(calificaciones)
desviacion_calificaciones = np.std(calificaciones)

# Imprimir los resultados
print(f"Media de calificaciones: {media_calificaciones}")
print(f"Desviación estándar de calificaciones: {desviacion_calificaciones}")

# Graficar el histograma de las calificaciones
plt.hist(calificaciones, bins=30, density=True, alpha=0.6, color='blue')

# Graficar la curva de la distribución normal teórica para las calificaciones
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, media_calificaciones, desviacion_calificaciones)
plt.plot(x, p, 'k', linewidth=2)

plt.title('Distribución de calificaciones simuladas')
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.show()

# Calcular la probabilidad de que un estudiante tenga una calificación superior a 85
probabilidad_mayor_85 = 1 - norm.cdf(85, media_calificaciones, desviacion_calificaciones)
print(f"Probabilidad de que un estudiante tenga una calificación superior a 85: {probabilidad_mayor_85:.4f}")
```
