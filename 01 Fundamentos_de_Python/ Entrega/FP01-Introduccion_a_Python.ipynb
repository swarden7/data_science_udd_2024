Entonces, nos surgió la inquietud sobre si podemos llevar esto un poco más lejos utilizando **emojis** en el código.

Al investigar un poco, encontramos que los Emojis [son en realidad letras o caracteres en codificados en formato UTF-8](https://www.w3schools.com/html/html_emojis.asp#:~:text=Emojis%20look%20like%20images%2C%20or,8%20(Unicode)%20character%20set.), por lo que no debiésemos tener problemas.

Probaremos creando una función que devuelve un emoji dependiendo de una emoción predefinida dentro de un conjunto de estas.

````python
def emoción2emoji(emoción:str):
    """
    Devuelve un emoji según un string predefinido.
    Las opciones son:
        - feliz
        - pokerface
        - triste
    """

    if emoción == 'feliz':
        return '😊'

    elif emoción == 'pokerface':
        return '😐'

    elif emoción == 'triste':
        return '😔'

    else:
        return '🤔'

emoción2emoji('feliz')
````

¡Funciona!

Si bien pueden ser pocas las aplicaciones de utilizar emojis en el código, podemos apreciar la versatilidad de Python al ser capaz de manejar sin problemas todo tipo de caracteres UTF-8. Ahora tener una mala ortografía no es la excusa para no utilizar tildes 🤭.

De igual forma es importante tomar en consideración cuáles son las guías de estilo internas de la organización en la que uno se desempeña, pues si bien 'se puede', si todo el codebase está sin tildes no tiene sentido ponerse a usar tildes en todo, pues puede llevar a bugs evitables.

¿Qué pasaría si quiero cambiar la codificación de mi notebook para que sea distinta a UTF-8?

Primero reviso todas las opciones que tengo de codificación.

````python
# Con la biblioteca encodings podemos revisar las codificaciones disponibles en nuestro entorno:
import encodings
import pkgutil

# Obtener todas las codificaciones disponibles
modulos_codificacion = set([mod.name for mod in pkgutil.iter_modules(encodings.__path__)])

# Mostrar las codificaciones
print(sorted(modulos_codificacion))
````

Con la función encoding puedo especificar con que codificación quiero abrir o guardar archivos. También, puedo codificar un string en una codificación específica. Esto convierte el string a una secuencia de bytes en una codificación específica.



````python
canción = "Huey Lewis And The News - If This Is It"
# Por forma predeterminada este string está en UTF-8.
# Definimos el mismo string con otra codificación
canción2 = "Huey Lewis And The News - If This Is It".encode('iso8859_1')
print(canción)
print(canción2)

# Se ve que son casi iguales excepto por el inicio. Hacemos un ejemplo con tildes y eñes:
texto = "El niño cantó una canción"
texto2 = "El niño cantó una canción".encode('iso8859_1')
print(texto)
print(texto2)
````

Probando distintas codificaciones vemos que "ascii" entrega error frente a tildes, eñes, emojis. Aun así, si es que definimos el nombre del objeto codificado en ascii con una tilde esto no genera error. \
En cambio, 'iso8859_1' puede procesar string con eñes y tildes, pero no identifica estos como tales al imprimirlo. Además, no es capaz de identificar emojis.

Si quiero decodificar un byte string que está en codificación 'iso8859_1' puedo utilizar la función decode. Veremos igualmente que si intento interpretarlo con una codificación utf-8 entrega error, hasta que especifico cómo tratar con los errores.



````python
# Decodificar con 'iso8859_1'
texto3 = texto2.decode('iso8859_1')
print(texto3)

## Decodificar con 'utf_8'
texto4 = texto2.decode('utf_8')
print(texto4)
## Esto lanza error
````

````python
# Decodificar con 'utf_8' especificando tratamiento de errores
texto4 = texto2.decode('utf_8', errors='ignore')
print(texto4)
# Vemos que las letras con tildes y eñe desaparecen
# También puedo reemplazarlas por '?'
texto4 = texto2.decode('utf_8', errors='replace')
print(texto4)
````

Como mencionamos anteriormente, también podemos abrir archivos con encode, en caso que sepamos que estos tienen una codificación específica.
